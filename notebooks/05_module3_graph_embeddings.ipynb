```python
# ============================================================
# MODULE 3 — GRAPH EMBEDDINGS SANS API EXTERNE
# VERSION : TITRE + TEXTE
#
# Contexte : Wikidata bloqué (403) depuis Colab.
# Solution : construire un graphe de co-occurrence d'entités
#            à partir du dataset lui-même, puis entraîner
#            TransE dessus avec PyKEEN.
#
# Principe :
#   - spaCy extrait les entités de chaque article
#   - On construit des triplets (entité_A, co-occur_in_article, entité_B)
#     pour toutes les paires d'entités dans le même article
#   - On enrichit avec des triplets de type (entité, type, PERSON/ORG/GPE...)
#   - TransE apprend des embeddings qui capturent ces co-occurrences
#   - C'est un vrai graph embedding, entraîné sur des relations réelles
#     du corpus — sans dépendance réseau
# ============================================================


# CELLULE 1 — INSTALLATION
!pip install spacy pykeen -q
!python -m spacy download en_core_web_sm -q
print("✅ Installations terminées")


# CELLULE 2 — IMPORTS
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel
from transformers import get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             roc_auc_score, classification_report)
from sklearn.utils.class_weight import compute_class_weight
import spacy, os, pickle, re
from urllib.parse import urlparse
from google.colab import drive
import matplotlib.pyplot as plt
from itertools import combinations

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Appareil utilisé : {device}")
nlp = spacy.load("en_core_web_sm")
print("✅ spaCy chargé")


# CELLULE 3 — MODULE 2 INTÉGRÉ (titre+texte, dim=5)
MBFC_CREDIBILITY_DB = {
    # --- Sources FIABLES (score élevé) ---
    "reuters.com":          {"score": 0.95, "bias": "center",       "category": "reliable"},
    "apnews.com":           {"score": 0.95, "bias": "center",       "category": "reliable"},
    "bbc.com":              {"score": 0.90, "bias": "center-left",  "category": "reliable"},
    "theguardian.com":      {"score": 0.85, "bias": "center-left",  "category": "reliable"},
    "nytimes.com":          {"score": 0.85, "bias": "center-left",  "category": "reliable"},
    "washingtonpost.com":   {"score": 0.85, "bias": "center-left",  "category": "reliable"},
    "npr.org":              {"score": 0.88, "bias": "center-left",  "category": "reliable"},
    "politifact.com":       {"score": 0.92, "bias": "center",       "category": "fact-checker"},
    "snopes.com":           {"score": 0.90, "bias": "center-left",  "category": "fact-checker"},
    "usatoday.com":         {"score": 0.80, "bias": "center",       "category": "reliable"},
    "cbsnews.com":          {"score": 0.82, "bias": "center-left",  "category": "reliable"},
    "nbcnews.com":          {"score": 0.80, "bias": "center-left",  "category": "reliable"},
    "abcnews.go.com":       {"score": 0.80, "bias": "center-left",  "category": "reliable"},

    # --- Sources MIXTES (score moyen) ---
    "foxnews.com":          {"score": 0.45, "bias": "right",        "category": "mixed"},
    "dailymail.co.uk":      {"score": 0.40, "bias": "right",        "category": "mixed"},
    "nypost.com":           {"score": 0.42, "bias": "right-center", "category": "mixed"},
    "huffpost.com":         {"score": 0.55, "bias": "left",         "category": "mixed"},
    "salon.com":            {"score": 0.50, "bias": "left",         "category": "mixed"},
    "thehill.com":          {"score": 0.65, "bias": "center",       "category": "mixed"},

    # --- Sources PEU FIABLES (score bas) ---
    "breitbart.com":        {"score": 0.10, "bias": "extreme-right","category": "fake"},
    "infowars.com":         {"score": 0.05, "bias": "extreme-right","category": "fake"},
    "naturalnews.com":      {"score": 0.05, "bias": "extreme-right","category": "fake"},
    "beforeitsnews.com":    {"score": 0.08, "bias": "conspiracy",   "category": "fake"},
    "worldnewsdailyreport.com": {"score": 0.02, "bias": "fake",    "category": "fake"},
    "empirenews.net":       {"score": 0.03, "bias": "fake",         "category": "fake"},
    "abcnews.com.co":       {"score": 0.02, "bias": "fake",         "category": "fake"},
    "politicot.com":        {"score": 0.05, "bias": "fake",         "category": "fake"},
    "newslo.com":           {"score": 0.08, "bias": "fake",         "category": "fake"},
    "thelastlineofdefense.org": {"score": 0.03, "bias": "fake",    "category": "fake"},
}
DEFAULT_INFO = {"score": 0.5, "bias": "unknown", "category": "unknown"}

SENSATIONAL_WORDS = [
    "shocking", "unbelievable", "breaking", "exclusive", "bombshell",
    "outrage", "scandal", "exposed", "revealed", "secret", "conspiracy",
    "urgent", "alert", "warning", "crisis", "disaster", "chaos",
    "incredible", "stunning", "explosive", "miracle", "banned",
    "censored", "truth", "wake up", "they don't want you to know"
]

def extract_domain(url_or_source):
    if not isinstance(url_or_source, str):
        return "unknown"
    try:
        url_or_source = url_or_source.strip()
        if url_or_source.startswith("http"):
            domain = urlparse(url_or_source).netloc.replace("www.", "")
            return domain if domain else "unknown"
        if "." in url_or_source:
            return url_or_source.replace("www.", "")
        return url_or_source
    except:
        return "unknown"

def get_mbfc_score(domain):
    domain = domain.lower().strip()
    if domain in MBFC_CREDIBILITY_DB:
        return MBFC_CREDIBILITY_DB[domain]
    for known, info in MBFC_CREDIBILITY_DB.items():
        if known in domain or domain in known:
            return info
    return DEFAULT_INFO

def compute_source_vector(domain_or_url, title, text=""):
    domain     = extract_domain(domain_or_url)
    mbfc_score = get_mbfc_score(domain)["score"]
    full       = title + " " + text
    nb_upper   = sum(1 for c in title if c.isupper())
    nb_letters = sum(1 for c in title if c.isalpha())
    ratio_up   = nb_upper / nb_letters if nb_letters > 0 else 0.0
    nb_excl    = full.count("!") + full.count("?")
    nb_words   = len(full.split())
    excl_sc    = nb_excl / nb_words if nb_words > 0 else 0.0
    nb_sens    = sum(1 for w in SENSATIONAL_WORDS if w in full.lower())
    sens_sc    = nb_sens / len(SENSATIONAL_WORDS)
    nb_links   = len(re.findall(r'http[s]?://\S+', text))
    return np.array([mbfc_score, ratio_up, excl_sc, sens_sc,
                     min(nb_links / 10, 1.0)], dtype=np.float32)

SOURCE_DIM = 5
print("✅ Module 2 chargé (dim=5)")


# CELLULE 4 — NER
def extract_entities(title, text=""):
    text_part  = ' '.join(str(text).split()[:150]) if text else ""
    full_input = f"{title} {text_part}".strip()
    doc        = nlp(full_input)
    types_ok   = {"PERSON", "ORG", "GPE", "LOC", "EVENT"}
    seen, result = set(), []
    for ent in doc.ents:
        if ent.label_ in types_ok:
            t = ent.text.strip().strip("'s").strip()
            if len(t) > 1 and t.lower() not in seen:
                seen.add(t.lower())
                result.append((t.lower(), ent.label_))  # normalisation en minuscules
    return result

print("✅ NER prêt")


# CELLULE 5 — CONSTRUCTION DU GRAPHE DE CO-OCCURRENCE
# ============================================================
# On construit deux types de triplets :
#
# Type A — Co-occurrence dans le même article :
#   (entité_A, "co_occurs_with", entité_B)
#   → capture quelles personnes/organisations apparaissent ensemble
#   → ex: ("kim kardashian", "co_occurs_with", "kanye west")
#
# Type B — Relation entité-type :
#   (entité, "has_type", TYPE_PERSON/ORG/GPE/LOC/EVENT)
#   → capture la catégorie sémantique de chaque entité
#   → ex: ("kim kardashian", "has_type", "TYPE_PERSON")
#
# Ce graphe est 100% local, sans API externe.
# Il capture les relations réelles dans le corpus GossipCop.
# ============================================================

from pykeen.pipeline import pipeline as pykeen_pipeline
from pykeen.triples import TriplesFactory

def build_cooccurrence_graph(titles, texts,
                              cache_path="cooccurrence_triplets.pkl",
                              min_entity_freq=2):
    """
    Construit le graphe de co-occurrence depuis le corpus.

    min_entity_freq : une entité doit apparaître dans au moins
                      N articles pour être incluse (filtre le bruit).
    """
    if os.path.exists(cache_path):
        print(f"✅ Triplets chargés depuis '{cache_path}'")
        with open(cache_path, 'rb') as f:
            return pickle.load(f)

    print("Construction du graphe de co-occurrence...")

    # Étape 1 : extraire toutes les entités avec leur fréquence
    entity_freq = {}
    all_article_entities = []

    for i, (title, text) in enumerate(zip(titles, texts)):
        ents = extract_entities(title, text)
        all_article_entities.append(ents)
        for ent, typ in ents:
            entity_freq[ent] = entity_freq.get(ent, 0) + 1
        if (i + 1) % 500 == 0:
            print(f"  NER : {i+1}/{len(titles)} articles")

    # Filtre : garder seulement les entités fréquentes
    frequent_entities = {e for e, f in entity_freq.items() if f >= min_entity_freq}
    print(f"   Entités totales    : {len(entity_freq)}")
    print(f"   Entités fréquentes (>={min_entity_freq}): {len(frequent_entities)}")

    # Étape 2 : construire les triplets
    triplets = set()

    for ents in all_article_entities:
        # Filtrer aux entités fréquentes
        ents_filtered = [(e, t) for e, t in ents if e in frequent_entities]

        # Type B : entité → type
        for ent, typ in ents_filtered:
            triplets.add((ent, "has_type", f"TYPE_{typ}"))

        # Type A : co-occurrence (toutes les paires)
        # On limite à 5 entités par article pour éviter explosion combinatoire
        ents_limited = ents_filtered[:5]
        for (e1, t1), (e2, t2) in combinations(ents_limited, 2):
            if e1 != e2:
                triplets.add((e1, "co_occurs_with", e2))
                triplets.add((e2, "co_occurs_with", e1))  # symétrique

    triplets = list(triplets)
    print(f"   Triplets générés   : {len(triplets)}")
    print(f"     dont co-occurrence : {sum(1 for t in triplets if t[1]=='co_occurs_with')}")
    print(f"     dont has_type      : {sum(1 for t in triplets if t[1]=='has_type')}")

    with open(cache_path, 'wb') as f:
        pickle.dump(triplets, f)

    return triplets


def train_transe_on_corpus(triplets,
                            embedding_dim  = 64,
                            num_epochs     = 50,
                            cache_path_emb = "transe_corpus_emb.pkl"):
    """
    Entraîne TransE sur le graphe de co-occurrence du corpus.
    Retourne un dict {entity_str: np.array(embedding_dim)}.
    """
    if os.path.exists(cache_path_emb):
        print(f"✅ Embeddings TransE chargés depuis '{cache_path_emb}'")
        with open(cache_path_emb, 'rb') as f:
            return pickle.load(f)

    if len(triplets) < 10:
        print("⚠️  Pas assez de triplets")
        return {}

    triples_array = np.array(triplets, dtype=str)
    tf = TriplesFactory.from_labeled_triples(triples_array)
    print(f"   Graphe : {tf.num_entities} entités | {tf.num_relations} relations | {len(triplets)} triplets")

    train_tf, test_tf = tf.split([0.8, 0.2], random_state=42)

    print(f"Entraînement TransE (dim={embedding_dim}, epochs={num_epochs})...")
    result = pykeen_pipeline(
        training         = train_tf,
        testing          = test_tf,
        model            = 'TransE',
        model_kwargs     = dict(embedding_dim=embedding_dim),
        training_kwargs  = dict(num_epochs=num_epochs, batch_size=256),
        optimizer        = 'Adam',
        optimizer_kwargs = dict(lr=0.01),
        random_seed      = 42,
        device           = 'cuda' if torch.cuda.is_available() else 'cpu',
    )
    print("✅ TransE entraîné")

    # Extraction des embeddings
    model_pk     = result.model
    entity_repr  = model_pk.entity_representations[0]
    all_emb      = entity_repr().detach().cpu().numpy()
    entity_to_id = tf.entity_to_id

    emb_dict = {e: all_emb[idx] for e, idx in entity_to_id.items()}

    with open(cache_path_emb, 'wb') as f:
        pickle.dump(emb_dict, f)
    print(f"✅ {len(emb_dict)} embeddings sauvegardés dans '{cache_path_emb}'")
    return emb_dict

print("✅ Fonctions graphe prêtes")


# CELLULE 6 — VECTEUR KG PAR ARTICLE
EMBEDDING_DIM = 64
KG_DIM        = EMBEDDING_DIM + 4   # 68

def get_kg_vector_for_article(title, text, emb_dict):
    """
    Calcule le vecteur KG d'un article via les embeddings TransE
    du graphe de co-occurrence.
    """
    entities = extract_entities(title, text)
    nb_found = len(entities)

    if nb_found == 0:
        return np.zeros(KG_DIM, dtype=np.float32)

    embeddings  = []
    nb_in_graph = 0
    for ent, _ in entities:
        if ent in emb_dict:
            embeddings.append(emb_dict[ent])
            nb_in_graph += 1
        else:
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))

    mean_emb = np.mean(embeddings, axis=0).astype(np.float32)
    norm = np.linalg.norm(mean_emb)
    if norm > 0:
        mean_emb = mean_emb / norm

    ratio_in_graph = nb_in_graph / nb_found if nb_found > 0 else 0.0
    coverage = np.array([
        min(nb_found    / 5.0, 1.0),
        ratio_in_graph,
        float(nb_in_graph > 0),
        min(nb_in_graph / 3.0, 1.0),
    ], dtype=np.float32)

    return np.concatenate([mean_emb, coverage])

print(f"✅ KG_DIM = {KG_DIM}  (TransE {EMBEDDING_DIM}d + 4 coverage features)")


# CELLULE 7 — CHARGEMENT DATASET + PRÉCALCUL
drive.mount('/content/drive')

DATASET_PATH    = '/content/drive/MyDrive/fakenewsnet_gossipcop_balanced.csv'
KG_CACHE_PATH   = '/content/drive/MyDrive/kg_vectors_cooccur_tt.npy'
TRANSE_EMB_PATH = '/content/drive/MyDrive/transe_cooccur_tt_emb.pkl'
TRIPLETS_PATH   = '/content/drive/MyDrive/cooccurrence_triplets_tt.pkl'

df = pd.read_csv(DATASET_PATH)
if 'domain' not in df.columns:
    df['domain'] = df['url'].apply(
        lambda u: urlparse(str(u)).netloc.replace('www.', '') or 'unknown')
df['title']  = df['title'].fillna('')
df['text']   = df['text'].fillna('') if 'text' in df.columns else ''
df['domain'] = df['domain'].fillna('unknown').replace('nan', 'unknown')
df['label']  = df['label'].astype(int)
df = df.dropna(subset=['label']).reset_index(drop=True)

titles  = df['title'].tolist()
texts   = df['text'].apply(lambda t: ' '.join(str(t).split()[:150])).tolist() \
          if 'text' in df.columns else [''] * len(df)
labels  = df['label'].tolist()
domains = df['domain'].tolist()
print(f"Dataset : {len(titles)} articles")

# Test NER sur quelques exemples réels
print("\nTest NER sur 5 articles du dataset :")
for i in range(min(5, len(titles))):
    ents = extract_entities(titles[i], texts[i])
    print(f"  [{i}] '{titles[i][:55]}' → {[e for e,_ in ents]}")

# ---- Étape 1 : Construction du graphe ----
print("\n[1/3] Construction du graphe de co-occurrence...")
triplets = build_cooccurrence_graph(
    titles, texts,
    cache_path     = TRIPLETS_PATH,
    min_entity_freq = 2   # entité doit apparaître dans ≥2 articles
)

# ---- Étape 2 : Entraînement TransE ----
print("\n[2/3] Entraînement TransE sur le graphe corpus...")
emb_dict = train_transe_on_corpus(
    triplets,
    embedding_dim  = EMBEDDING_DIM,
    num_epochs     = 50,
    cache_path_emb = TRANSE_EMB_PATH,
)
print(f"   Entités dans le graphe : {len(emb_dict)}")

# ---- Étape 3 : Vecteurs KG par article ----
print("\n[3/3] Précalcul des vecteurs KG par article...")
if os.path.exists(KG_CACHE_PATH):
    kg_matrix = np.load(KG_CACHE_PATH)
    print(f"✅ KG vectors chargés : {kg_matrix.shape}")
else:
    kg_vectors = []
    for i, (title, text) in enumerate(zip(titles, texts)):
        kg_vectors.append(get_kg_vector_for_article(title, text, emb_dict))
        if (i + 1) % 500 == 0:
            print(f"  {i+1}/{len(titles)}...")
    kg_matrix = np.vstack(kg_vectors)
    np.save(KG_CACHE_PATH, kg_matrix)
    print(f"✅ KG vectors calculés : {kg_matrix.shape}")

# ---- Vecteurs sources ----
print("Calcul vecteurs sources (Module 2)...")
source_vectors = np.vstack([
    compute_source_vector(d, t, x)
    for d, t, x in zip(domains, titles, texts)
])
print(f"✅ Source vectors : {source_vectors.shape}")

# Diagnostic qualité KG
non_zero = np.sum(np.any(kg_matrix != 0, axis=1))
print(f"\nArticles avec KG non-nul : {non_zero}/{len(titles)} ({non_zero/len(titles)*100:.1f}%)")

print(f"\n{'='*50}\nRÉCAPITULATIF\n{'='*50}")
print(f"Articles       : {len(titles)}")
print(f"Source vectors : {source_vectors.shape}  ← Module 2")
print(f"KG vectors     : {kg_matrix.shape}  ← Module 3 (TransE co-occurrence)")
print(f"Entités graph  : {len(emb_dict)}")
print(f"Triplets graph : {len(triplets)}")


# CELLULE 8 — DATASET PYTORCH TRIMODAL
class TrimodalDataset(Dataset):
    def __init__(self, texts, source_vectors, kg_vectors,
                 labels, tokenizer, max_length=256):
        self.texts          = texts
        self.source_vectors = source_vectors
        self.kg_vectors     = kg_vectors
        self.labels         = labels
        self.tokenizer      = tokenizer
        self.max_length     = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            str(self.texts[idx]),
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids':      enc['input_ids'].squeeze(),
            'attention_mask': enc['attention_mask'].squeeze(),
            'source_vector':  torch.tensor(self.source_vectors[idx], dtype=torch.float32),
            'kg_vector':      torch.tensor(self.kg_vectors[idx],     dtype=torch.float32),
            'label':          torch.tensor(self.labels[idx],         dtype=torch.long)
        }


# CELLULE 9 — MODÈLE TRIMODAL
class TrimodalFakeNewsDetector(nn.Module):
    def __init__(self, bert_model_name='bert-base-uncased',
                 source_dim=5, kg_dim=68, hidden_dim=256,
                 num_classes=2, dropout=0.3):
        super().__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        bert_hidden = self.bert.config.hidden_size   # 768

        self.source_encoder = nn.Sequential(
            nn.Linear(source_dim, 32), nn.ReLU(),
            nn.Dropout(dropout), nn.Linear(32, 32), nn.ReLU()
        )
        self.kg_encoder = nn.Sequential(
            nn.Linear(kg_dim, 64), nn.ReLU(),
            nn.Dropout(dropout), nn.Linear(64, 64), nn.ReLU()
        )
        # 768 + 32 + 64 = 864
        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + 32 + 64, hidden_dim),
            nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(), nn.Dropout(dropout / 2),
            nn.Linear(64, num_classes)
        )

    def forward(self, input_ids, attention_mask, source_vector, kg_vector):
        bert_out = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls   = bert_out.last_hidden_state[:, 0, :]
        src   = self.source_encoder(source_vector)
        kg    = self.kg_encoder(kg_vector)
        fused = torch.cat([cls, src, kg], dim=1)
        return self.classifier(fused)


# CELLULE 10 — ENTRAÎNEMENT ET ÉVALUATION
def train_epoch(model, loader, optimizer, scheduler, criterion, device):
    model.train()
    total_loss, all_preds, all_lbls = 0, [], []
    for batch in loader:
        ids  = batch['input_ids'].to(device)
        mask = batch['attention_mask'].to(device)
        src  = batch['source_vector'].to(device)
        kg   = batch['kg_vector'].to(device)
        lbls = batch['label'].to(device)
        optimizer.zero_grad()
        logits = model(ids, mask, src, kg)
        loss   = criterion(logits, lbls)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()
        all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
        all_lbls.extend(lbls.cpu().numpy())
    return total_loss / len(loader), accuracy_score(all_lbls, all_preds)

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, all_preds, all_lbls, all_probs = 0, [], [], []
    with torch.no_grad():
        for batch in loader:
            ids  = batch['input_ids'].to(device)
            mask = batch['attention_mask'].to(device)
            src  = batch['source_vector'].to(device)
            kg   = batch['kg_vector'].to(device)
            lbls = batch['label'].to(device)
            logits = model(ids, mask, src, kg)
            loss   = criterion(logits, lbls)
            total_loss += loss.item()
            all_probs.extend(torch.softmax(logits, 1)[:, 1].cpu().numpy())
            all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
            all_lbls.extend(lbls.cpu().numpy())
    acc = accuracy_score(all_lbls, all_preds)
    f1  = f1_score(all_lbls, all_preds, average='weighted')
    auc = roc_auc_score(all_lbls, all_probs)
    return total_loss / len(loader), acc, f1, auc, all_preds, all_lbls


# CELLULE 11 — PIPELINE PRINCIPALE
MAX_LENGTH    = 256
BATCH_SIZE    = 8
EPOCHS        = 3
LEARNING_RATE = 2e-5

contents    = [f"{t} {x}".strip() for t, x in zip(titles, texts)]
bert_inputs = contents

(train_texts, test_texts,
 train_src,   test_src,
 train_kg,    test_kg,
 train_labels, test_labels) = train_test_split(
    bert_inputs, source_vectors, kg_matrix, labels,
    test_size=0.2, random_state=42, stratify=labels
)
print(f"Train : {len(train_texts)} | Test : {len(test_texts)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

train_ds = TrimodalDataset(train_texts, train_src, train_kg, train_labels, tokenizer, MAX_LENGTH)
test_ds  = TrimodalDataset(test_texts,  test_src,  test_kg,  test_labels,  tokenizer, MAX_LENGTH)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=2, pin_memory=True)

model = TrimodalFakeNewsDetector(source_dim=SOURCE_DIM, kg_dim=KG_DIM).to(device)
print(f"Paramètres totaux : {sum(p.numel() for p in model.parameters()):,}")

class_weights = compute_class_weight('balanced', classes=np.array([0,1]), y=train_labels)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor(class_weights, dtype=torch.float32).to(device))
optimizer   = AdamW(model.parameters(), lr=LEARNING_RATE)
total_steps = len(train_loader) * EPOCHS
scheduler   = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps   = total_steps // 10,
    num_training_steps = total_steps
)

print("\n=== ENTRAÎNEMENT M1+M2+M3 (TransE co-occurrence — Titre+Texte) ===\n")
history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
best_f1 = 0

for epoch in range(EPOCHS):
    print(f"--- Époque {epoch+1}/{EPOCHS} ---")
    tr_loss, tr_acc = train_epoch(model, train_loader, optimizer, scheduler, criterion, device)
    vl_loss, vl_acc, vl_f1, vl_auc, _, _ = evaluate(model, test_loader, criterion, device)
    history['train_loss'].append(tr_loss)
    history['train_acc'].append(tr_acc)
    history['val_loss'].append(vl_loss)
    history['val_acc'].append(vl_acc)
    print(f"Train → Loss: {tr_loss:.4f} | Acc: {tr_acc:.4f}")
    print(f"Test  → Loss: {vl_loss:.4f} | Acc: {vl_acc:.4f} | F1: {vl_f1:.4f} | AUC: {vl_auc:.4f}")
    if vl_f1 > best_f1:
        best_f1 = vl_f1
        torch.save(model.state_dict(), 'best_trimodal_cooccur_tt.pt')
        print(f"  ✅ Meilleur modèle (F1={best_f1:.4f})\n")
    else:
        print()

model.load_state_dict(torch.load('best_trimodal_cooccur_tt.pt'))
print("\n=== ÉVALUATION FINALE (M1+M2+M3 TransE — Titre+Texte) ===\n")
_, acc, f1, auc, preds, true_labels = evaluate(model, test_loader, criterion, device)
print(classification_report(true_labels, preds, target_names=['Réel', 'Fake']))
print(f"Accuracy : {acc:.4f} | F1 : {f1:.4f} | AUC : {auc:.4f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(history['train_loss'], label='Train', marker='o')
ax1.plot(history['val_loss'],   label='Val',   marker='o')
ax1.set_title('Loss — M1+M2+M3 TransE (Titre+Texte)') ; ax1.legend()
ax2.plot(history['train_acc'], label='Train', marker='o')
ax2.plot(history['val_acc'],   label='Val',   marker='o')
ax2.set_title('Accuracy — M1+M2+M3 TransE (Titre+Texte)') ; ax2.legend()
plt.tight_layout()
plt.savefig('trimodal_cooccur_tt_curves.png', dpi=120)
plt.show()

print("\n" + "="*65)
print("TABLEAU COMPARATIF — GossipCop (Titre+Texte)")
print("="*65)
print(f"{'Modèle':<42} {'Accuracy':>10} {'F1':>8} {'AUC':>8}")
print("-"*65)
print(f"{'TF-IDF+SVM (titre seul)':<42} {'0.7876':>10} {'0.7875':>8} {'0.8599':>8}")
print(f"{'TF-IDF+SVM (titre+texte)':<42} {'0.8293':>10} {'0.8293':>8} {'0.9066':>8}")
print(f"{'BERT+Source M1+M2 (titre seul)':<42} {'0.8235':>10} {'0.8234':>8} {'0.8942':>8}")
print(f"{'M1+M2+M3 TransE (titre+texte)':<42} {acc:>10.4f} {f1:>8.4f} {auc:>8.4f}")
print("="*65)

torch.save(model.state_dict(), 'trimodal_cooccur_tt_final.pt')
print("✅ Modèle final sauvegardé dans 'trimodal_cooccur_tt_final.pt'")
```

```python
# ============================================================
# MODULE 3 — GRAPH EMBEDDINGS SANS API EXTERNE
# VERSION : TITRE SEUL
# ============================================================


# CELLULE 1 — INSTALLATION
!pip install spacy pykeen -q
!python -m spacy download en_core_web_sm -q
print("✅ Installations terminées")


# CELLULE 2 — IMPORTS
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel
from transformers import get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             roc_auc_score, classification_report)
from sklearn.utils.class_weight import compute_class_weight
import spacy, os, pickle, re
from urllib.parse import urlparse
from google.colab import drive
import matplotlib.pyplot as plt
from itertools import combinations

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Appareil utilisé : {device}")
nlp = spacy.load("en_core_web_sm")
print("✅ spaCy chargé")


# CELLULE 3 — MODULE 2 INTÉGRÉ (titre seul, dim=4) ✅ TITRE SEUL
MBFC_CREDIBILITY_DB = {
    "reuters.com":          {"score": 0.95, "bias": "center",        "category": "reliable"},
    "apnews.com":           {"score": 0.95, "bias": "center",        "category": "reliable"},
    "bbc.com":              {"score": 0.90, "bias": "center-left",   "category": "reliable"},
    "theguardian.com":      {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "nytimes.com":          {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "washingtonpost.com":   {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "npr.org":              {"score": 0.88, "bias": "center-left",   "category": "reliable"},
    "politifact.com":       {"score": 0.92, "bias": "center",        "category": "fact-checker"},
    "snopes.com":           {"score": 0.90, "bias": "center-left",   "category": "fact-checker"},
    "foxnews.com":          {"score": 0.45, "bias": "right",         "category": "mixed"},
    "dailymail.co.uk":      {"score": 0.40, "bias": "right",         "category": "mixed"},
    "breitbart.com":        {"score": 0.10, "bias": "extreme-right", "category": "fake"},
    "infowars.com":         {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "naturalnews.com":      {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "worldnewsdailyreport.com": {"score": 0.02, "bias": "fake",     "category": "fake"},
}
DEFAULT_INFO = {"score": 0.5, "bias": "unknown", "category": "unknown"}

SENSATIONAL_WORDS = [
    "shocking", "unbelievable", "breaking", "exclusive", "bombshell",
    "outrage", "scandal", "exposed", "revealed", "secret", "conspiracy",
    "urgent", "alert", "warning", "crisis", "disaster", "chaos",
    "incredible", "stunning", "explosive", "miracle", "banned",
    "censored", "truth", "wake up", "they don't want you to know"
]

def extract_domain(url_or_source):
    if not isinstance(url_or_source, str):
        return "unknown"
    try:
        url_or_source = url_or_source.strip()
        if url_or_source.startswith("http"):
            domain = urlparse(url_or_source).netloc.replace("www.", "")
            return domain if domain else "unknown"
        if "." in url_or_source:
            return url_or_source.replace("www.", "")
        return url_or_source
    except:
        return "unknown"

def get_mbfc_score(domain):
    domain = domain.lower().strip()
    if domain in MBFC_CREDIBILITY_DB:
        return MBFC_CREDIBILITY_DB[domain]
    for known, info in MBFC_CREDIBILITY_DB.items():
        if known in domain or domain in known:
            return info
    return DEFAULT_INFO

def compute_source_vector(domain_or_url, title):
    # ✅ TITRE SEUL — pas de paramètre text, vecteur [4]
    domain     = extract_domain(domain_or_url)
    mbfc_score = get_mbfc_score(domain)["score"]
    nb_upper   = sum(1 for c in title if c.isupper())
    nb_letters = sum(1 for c in title if c.isalpha())
    ratio_up   = nb_upper / nb_letters if nb_letters > 0 else 0.0
    nb_excl    = title.count("!") + title.count("?")
    nb_words   = len(title.split())
    excl_sc    = nb_excl / nb_words if nb_words > 0 else 0.0
    nb_sens    = sum(1 for w in SENSATIONAL_WORDS if w in title.lower())
    sens_sc    = nb_sens / len(SENSATIONAL_WORDS)
    return np.array([mbfc_score, ratio_up, excl_sc, sens_sc], dtype=np.float32)

SOURCE_DIM = 4  # ✅ TITRE SEUL
print("✅ Module 2 chargé (dim=4)")


# CELLULE 4 — NER ✅ TITRE SEUL
def extract_entities(title):
    # ✅ TITRE SEUL — pas de paramètre text
    doc      = nlp(str(title))
    types_ok = {"PERSON", "ORG", "GPE", "LOC", "EVENT"}
    seen, result = set(), []
    for ent in doc.ents:
        if ent.label_ in types_ok:
            t = ent.text.strip().strip("'s").strip()
            if len(t) > 1 and t.lower() not in seen:
                seen.add(t.lower())
                result.append((t.lower(), ent.label_))
    return result

print("✅ NER prêt")


# CELLULE 5 — CONSTRUCTION DU GRAPHE DE CO-OCCURRENCE
from pykeen.pipeline import pipeline as pykeen_pipeline
from pykeen.triples import TriplesFactory

def build_cooccurrence_graph(titles,
                              cache_path="cooccurrence_triplets.pkl",
                              min_entity_freq=2):
    # ✅ TITRE SEUL — itère sur titles uniquement
    if os.path.exists(cache_path):
        print(f"✅ Triplets chargés depuis '{cache_path}'")
        with open(cache_path, 'rb') as f:
            return pickle.load(f)

    print("Construction du graphe de co-occurrence...")

    entity_freq = {}
    all_article_entities = []

    for i, title in enumerate(titles):
        ents = extract_entities(title)
        all_article_entities.append(ents)
        for ent, typ in ents:
            entity_freq[ent] = entity_freq.get(ent, 0) + 1
        if (i + 1) % 500 == 0:
            print(f"  NER : {i+1}/{len(titles)} articles")

    frequent_entities = {e for e, f in entity_freq.items() if f >= min_entity_freq}
    print(f"   Entités totales    : {len(entity_freq)}")
    print(f"   Entités fréquentes (>={min_entity_freq}): {len(frequent_entities)}")

    triplets = set()
    for ents in all_article_entities:
        ents_filtered = [(e, t) for e, t in ents if e in frequent_entities]
        for ent, typ in ents_filtered:
            triplets.add((ent, "has_type", f"TYPE_{typ}"))
        ents_limited = ents_filtered[:5]
        for (e1, t1), (e2, t2) in combinations(ents_limited, 2):
            if e1 != e2:
                triplets.add((e1, "co_occurs_with", e2))
                triplets.add((e2, "co_occurs_with", e1))

    triplets = list(triplets)
    print(f"   Triplets générés   : {len(triplets)}")
    print(f"     dont co-occurrence : {sum(1 for t in triplets if t[1]=='co_occurs_with')}")
    print(f"     dont has_type      : {sum(1 for t in triplets if t[1]=='has_type')}")

    with open(cache_path, 'wb') as f:
        pickle.dump(triplets, f)
    return triplets


def train_transe_on_corpus(triplets,
                            embedding_dim  = 64,
                            num_epochs     = 50,
                            cache_path_emb = "transe_corpus_emb.pkl"):
    if os.path.exists(cache_path_emb):
        print(f"✅ Embeddings TransE chargés depuis '{cache_path_emb}'")
        with open(cache_path_emb, 'rb') as f:
            return pickle.load(f)

    if len(triplets) < 10:
        print("⚠️  Pas assez de triplets")
        return {}

    triples_array = np.array(triplets, dtype=str)
    tf = TriplesFactory.from_labeled_triples(triples_array)
    print(f"   Graphe : {tf.num_entities} entités | {tf.num_relations} relations | {len(triplets)} triplets")

    train_tf, test_tf = tf.split([0.8, 0.2], random_state=42)

    print(f"Entraînement TransE (dim={embedding_dim}, epochs={num_epochs})...")
    result = pykeen_pipeline(
        training         = train_tf,
        testing          = test_tf,
        model            = 'TransE',
        model_kwargs     = dict(embedding_dim=embedding_dim),
        training_kwargs  = dict(num_epochs=num_epochs, batch_size=256),
        optimizer        = 'Adam',
        optimizer_kwargs = dict(lr=0.01),
        random_seed      = 42,
        device           = 'cuda' if torch.cuda.is_available() else 'cpu',
    )
    print("✅ TransE entraîné")

    model_pk     = result.model
    entity_repr  = model_pk.entity_representations[0]
    all_emb      = entity_repr().detach().cpu().numpy()
    entity_to_id = tf.entity_to_id
    emb_dict = {e: all_emb[idx] for e, idx in entity_to_id.items()}

    with open(cache_path_emb, 'wb') as f:
        pickle.dump(emb_dict, f)
    print(f"✅ {len(emb_dict)} embeddings sauvegardés dans '{cache_path_emb}'")
    return emb_dict

print("✅ Fonctions graphe prêtes")


# CELLULE 6 — VECTEUR KG PAR ARTICLE
EMBEDDING_DIM = 64
KG_DIM        = EMBEDDING_DIM + 4   # 68

def get_kg_vector_for_article(title, emb_dict):
    # ✅ TITRE SEUL — pas de paramètre text
    entities = extract_entities(title)
    nb_found = len(entities)

    if nb_found == 0:
        return np.zeros(KG_DIM, dtype=np.float32)

    embeddings  = []
    nb_in_graph = 0
    for ent, _ in entities:
        if ent in emb_dict:
            embeddings.append(emb_dict[ent])
            nb_in_graph += 1
        else:
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))

    mean_emb = np.mean(embeddings, axis=0).astype(np.float32)
    norm = np.linalg.norm(mean_emb)
    if norm > 0:
        mean_emb = mean_emb / norm

    ratio_in_graph = nb_in_graph / nb_found if nb_found > 0 else 0.0
    coverage = np.array([
        min(nb_found    / 5.0, 1.0),
        ratio_in_graph,
        float(nb_in_graph > 0),
        min(nb_in_graph / 3.0, 1.0),
    ], dtype=np.float32)

    return np.concatenate([mean_emb, coverage])

print(f"✅ KG_DIM = {KG_DIM}  (TransE {EMBEDDING_DIM}d + 4 coverage features)")


# CELLULE 7 — CHARGEMENT DATASET + PRÉCALCUL
drive.mount('/content/drive')

DATASET_PATH    = '/content/drive/MyDrive/fakenewsnet_gossipcop_balanced.csv'
KG_CACHE_PATH   = '/content/drive/MyDrive/kg_vectors_cooccur_title.npy'       # ✅ nom distinct
TRANSE_EMB_PATH = '/content/drive/MyDrive/transe_cooccur_title_emb.pkl'       # ✅ nom distinct
TRIPLETS_PATH   = '/content/drive/MyDrive/cooccurrence_triplets_title.pkl'    # ✅ nom distinct

df = pd.read_csv(DATASET_PATH)
if 'domain' not in df.columns:
    df['domain'] = df['url'].apply(
        lambda u: urlparse(str(u)).netloc.replace('www.', '') or 'unknown')
df['title']  = df['title'].fillna('')
df['domain'] = df['domain'].fillna('unknown').replace('nan', 'unknown')
df['label']  = df['label'].astype(int)
df = df.dropna(subset=['label']).reset_index(drop=True)

titles  = df['title'].tolist()
labels  = df['label'].tolist()
domains = df['domain'].tolist()
print(f"Dataset : {len(titles)} articles")

# Test NER
print("\nTest NER sur 5 articles du dataset :")
for i in range(min(5, len(titles))):
    ents = extract_entities(titles[i])
    print(f"  [{i}] '{titles[i][:55]}' → {[e for e,_ in ents]}")

# ---- Étape 1 : Construction du graphe ----
print("\n[1/3] Construction du graphe de co-occurrence...")
triplets = build_cooccurrence_graph(
    titles,
    cache_path      = TRIPLETS_PATH,
    min_entity_freq = 2
)

# ---- Étape 2 : Entraînement TransE ----
print("\n[2/3] Entraînement TransE sur le graphe corpus...")
emb_dict = train_transe_on_corpus(
    triplets,
    embedding_dim  = EMBEDDING_DIM,
    num_epochs     = 50,
    cache_path_emb = TRANSE_EMB_PATH,
)
print(f"   Entités dans le graphe : {len(emb_dict)}")

# ---- Étape 3 : Vecteurs KG par article ----
print("\n[3/3] Précalcul des vecteurs KG par article...")
if os.path.exists(KG_CACHE_PATH):
    kg_matrix = np.load(KG_CACHE_PATH)
    print(f"✅ KG vectors chargés : {kg_matrix.shape}")
else:
    kg_vectors = []
    for i, title in enumerate(titles):
        kg_vectors.append(get_kg_vector_for_article(title, emb_dict))
        if (i + 1) % 500 == 0:
            print(f"  {i+1}/{len(titles)}...")
    kg_matrix = np.vstack(kg_vectors)
    np.save(KG_CACHE_PATH, kg_matrix)
    print(f"✅ KG vectors calculés : {kg_matrix.shape}")

# ---- Vecteurs sources ----
print("Calcul vecteurs sources (Module 2)...")
source_vectors = np.vstack([
    compute_source_vector(d, t)
    for d, t in zip(domains, titles)
])
print(f"✅ Source vectors : {source_vectors.shape}")  # (N, 4)

non_zero = np.sum(np.any(kg_matrix != 0, axis=1))
print(f"\nArticles avec KG non-nul : {non_zero}/{len(titles)} ({non_zero/len(titles)*100:.1f}%)")

print(f"\n{'='*50}\nRÉCAPITULATIF\n{'='*50}")
print(f"Articles       : {len(titles)}")
print(f"Source vectors : {source_vectors.shape}  ← Module 2")
print(f"KG vectors     : {kg_matrix.shape}  ← Module 3 (TransE co-occurrence)")
print(f"Entités graph  : {len(emb_dict)}")
print(f"Triplets graph : {len(triplets)}")


# CELLULE 8 — DATASET PYTORCH TRIMODAL
class TrimodalDataset(Dataset):
    def __init__(self, texts, source_vectors, kg_vectors,
                 labels, tokenizer, max_length=128):  # ✅ TITRE SEUL — max_length=128
        self.texts          = texts
        self.source_vectors = source_vectors
        self.kg_vectors     = kg_vectors
        self.labels         = labels
        self.tokenizer      = tokenizer
        self.max_length     = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            str(self.texts[idx]),
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids':      enc['input_ids'].squeeze(),
            'attention_mask': enc['attention_mask'].squeeze(),
            'source_vector':  torch.tensor(self.source_vectors[idx], dtype=torch.float32),
            'kg_vector':      torch.tensor(self.kg_vectors[idx],     dtype=torch.float32),
            'label':          torch.tensor(self.labels[idx],         dtype=torch.long)
        }


# CELLULE 9 — MODÈLE TRIMODAL
class TrimodalFakeNewsDetector(nn.Module):
    def __init__(self, bert_model_name='bert-base-uncased',
                 source_dim=4,   # ✅ TITRE SEUL
                 kg_dim=68, hidden_dim=256,
                 num_classes=2, dropout=0.3):
        super().__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        bert_hidden = self.bert.config.hidden_size   # 768

        self.source_encoder = nn.Sequential(
            nn.Linear(source_dim, 32), nn.ReLU(),
            nn.Dropout(dropout), nn.Linear(32, 32), nn.ReLU()
        )
        self.kg_encoder = nn.Sequential(
            nn.Linear(kg_dim, 64), nn.ReLU(),
            nn.Dropout(dropout), nn.Linear(64, 64), nn.ReLU()
        )
        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + 32 + 64, hidden_dim),
            nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(), nn.Dropout(dropout / 2),
            nn.Linear(64, num_classes)
        )

    def forward(self, input_ids, attention_mask, source_vector, kg_vector):
        bert_out = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls   = bert_out.last_hidden_state[:, 0, :]
        src   = self.source_encoder(source_vector)
        kg    = self.kg_encoder(kg_vector)
        fused = torch.cat([cls, src, kg], dim=1)
        return self.classifier(fused)


# CELLULE 10 — ENTRAÎNEMENT ET ÉVALUATION
def train_epoch(model, loader, optimizer, scheduler, criterion, device):
    model.train()
    total_loss, all_preds, all_lbls = 0, [], []
    for batch in loader:
        ids  = batch['input_ids'].to(device)
        mask = batch['attention_mask'].to(device)
        src  = batch['source_vector'].to(device)
        kg   = batch['kg_vector'].to(device)
        lbls = batch['label'].to(device)
        optimizer.zero_grad()
        logits = model(ids, mask, src, kg)
        loss   = criterion(logits, lbls)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()
        all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
        all_lbls.extend(lbls.cpu().numpy())
    return total_loss / len(loader), accuracy_score(all_lbls, all_preds)

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, all_preds, all_lbls, all_probs = 0, [], [], []
    with torch.no_grad():
        for batch in loader:
            ids  = batch['input_ids'].to(device)
            mask = batch['attention_mask'].to(device)
            src  = batch['source_vector'].to(device)
            kg   = batch['kg_vector'].to(device)
            lbls = batch['label'].to(device)
            logits = model(ids, mask, src, kg)
            loss   = criterion(logits, lbls)
            total_loss += loss.item()
            all_probs.extend(torch.softmax(logits, 1)[:, 1].cpu().numpy())
            all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
            all_lbls.extend(lbls.cpu().numpy())
    acc = accuracy_score(all_lbls, all_preds)
    f1  = f1_score(all_lbls, all_preds, average='weighted')
    auc = roc_auc_score(all_lbls, all_probs)
    return total_loss / len(loader), acc, f1, auc, all_preds, all_lbls


# CELLULE 11 — PIPELINE PRINCIPALE
MAX_LENGTH    = 128   # ✅ TITRE SEUL
BATCH_SIZE    = 16    # ✅ plus grand batch possible grâce à séquences plus courtes
EPOCHS        = 3
LEARNING_RATE = 2e-5

# ✅ TITRE SEUL — BERT reçoit uniquement le titre
bert_inputs = titles

(train_texts, test_texts,
 train_src,   test_src,
 train_kg,    test_kg,
 train_labels, test_labels) = train_test_split(
    bert_inputs, source_vectors, kg_matrix, labels,
    test_size=0.2, random_state=42, stratify=labels
)
print(f"Train : {len(train_texts)} | Test : {len(test_texts)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

train_ds = TrimodalDataset(train_texts, train_src, train_kg, train_labels, tokenizer, MAX_LENGTH)
test_ds  = TrimodalDataset(test_texts,  test_src,  test_kg,  test_labels,  tokenizer, MAX_LENGTH)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=2, pin_memory=True)

model = TrimodalFakeNewsDetector(source_dim=SOURCE_DIM, kg_dim=KG_DIM).to(device)
print(f"Paramètres totaux : {sum(p.numel() for p in model.parameters()):,}")

class_weights = compute_class_weight('balanced', classes=np.array([0,1]), y=train_labels)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor(class_weights, dtype=torch.float32).to(device))
optimizer   = AdamW(model.parameters(), lr=LEARNING_RATE)
total_steps = len(train_loader) * EPOCHS
scheduler   = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps   = total_steps // 10,
    num_training_steps = total_steps
)

print("\n=== ENTRAÎNEMENT M1+M2+M3 (TransE co-occurrence — Titre seul) ===\n")
history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
best_f1 = 0

for epoch in range(EPOCHS):
    print(f"--- Époque {epoch+1}/{EPOCHS} ---")
    tr_loss, tr_acc = train_epoch(model, train_loader, optimizer, scheduler, criterion, device)
    vl_loss, vl_acc, vl_f1, vl_auc, _, _ = evaluate(model, test_loader, criterion, device)
    history['train_loss'].append(tr_loss)
    history['train_acc'].append(tr_acc)
    history['val_loss'].append(vl_loss)
    history['val_acc'].append(vl_acc)
    print(f"Train → Loss: {tr_loss:.4f} | Acc: {tr_acc:.4f}")
    print(f"Test  → Loss: {vl_loss:.4f} | Acc: {vl_acc:.4f} | F1: {vl_f1:.4f} | AUC: {vl_auc:.4f}")
    if vl_f1 > best_f1:
        best_f1 = vl_f1
        torch.save(model.state_dict(), 'best_trimodal_cooccur_title.pt')
        print(f"  ✅ Meilleur modèle (F1={best_f1:.4f})\n")
    else:
        print()

model.load_state_dict(torch.load('best_trimodal_cooccur_title.pt'))
print("\n=== ÉVALUATION FINALE (M1+M2+M3 TransE — Titre seul) ===\n")
_, acc, f1, auc, preds, true_labels = evaluate(model, test_loader, criterion, device)
print(classification_report(true_labels, preds, target_names=['Réel', 'Fake']))
print(f"Accuracy : {acc:.4f} | F1 : {f1:.4f} | AUC : {auc:.4f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(history['train_loss'], label='Train', marker='o')
ax1.plot(history['val_loss'],   label='Val',   marker='o')
ax1.set_title('Loss — M1+M2+M3 TransE (Titre seul)') ; ax1.legend()
ax2.plot(history['train_acc'], label='Train', marker='o')
ax2.plot(history['val_acc'],   label='Val',   marker='o')
ax2.set_title('Accuracy — M1+M2+M3 TransE (Titre seul)') ; ax2.legend()
plt.tight_layout()
plt.savefig('trimodal_cooccur_title_curves.png', dpi=120)
plt.show()

print("\n" + "="*65)
print("TABLEAU COMPARATIF — GossipCop (Titre seul)")
print("="*65)
print(f"{'Modèle':<42} {'Accuracy':>10} {'F1':>8} {'AUC':>8}")
print("-"*65)
print(f"{'TF-IDF+SVM (titre seul)':<42} {'0.7876':>10} {'0.7875':>8} {'0.8599':>8}")
print(f"{'TF-IDF+SVM (titre+texte)':<42} {'0.8293':>10} {'0.8293':>8} {'0.9066':>8}")
print(f"{'BERT+Source M1+M2 (titre seul)':<42} {'0.8235':>10} {'0.8234':>8} {'0.8942':>8}")
print(f"{'M1+M2+M3 TransE tt (titre+texte)':<42} {'0.8147':>10} {'0.8147':>8} {'0.8943':>8}")
print(f"{'M1+M2+M3 TransE (titre seul)':<42} {acc:>10.4f} {f1:>8.4f} {auc:>8.4f}")
print("="*65)

torch.save(model.state_dict(), 'trimodal_cooccur_title_final.pt')
print("✅ Modèle final sauvegardé dans 'trimodal_cooccur_title_final.pt'")
```

```python
# ============================================================
# MODULE 3 — VRAIS GRAPH EMBEDDINGS (TransE via PyKEEN)
# Approche : sous-graphe Wikidata local → entraînement TransE
# Compatible : titre seul (source_dim=4) ou titre+texte (source_dim=5)
# ============================================================

# ============================================================
# CELLULE 1 — INSTALLATION
# ============================================================
!pip install spacy pykeen requests -q
!python -m spacy download en_core_web_sm -q
print("✅ Installations terminées")


# ============================================================
# CELLULE 2 — IMPORTS
# ============================================================
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel
from transformers import get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             roc_auc_score, classification_report)
from sklearn.utils.class_weight import compute_class_weight
import spacy, requests, time, os, pickle, re
from urllib.parse import urlparse
from google.colab import drive
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Appareil utilisé : {device}")
nlp = spacy.load("en_core_web_sm")
print("✅ spaCy chargé")


# ============================================================
# CELLULE 3 — MODULE 2 INTÉGRÉ (titre seul, dim=4)
# Remplace source_dim=5 et les deux dernières lignes si tu veux
# la version titre+texte — le reste du code ne change pas.
# ============================================================

MBFC_CREDIBILITY_DB = {
    "reuters.com":          {"score": 0.95, "bias": "center",        "category": "reliable"},
    "apnews.com":           {"score": 0.95, "bias": "center",        "category": "reliable"},
    "bbc.com":              {"score": 0.90, "bias": "center-left",   "category": "reliable"},
    "theguardian.com":      {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "nytimes.com":          {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "washingtonpost.com":   {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "npr.org":              {"score": 0.88, "bias": "center-left",   "category": "reliable"},
    "politifact.com":       {"score": 0.92, "bias": "center",        "category": "fact-checker"},
    "snopes.com":           {"score": 0.90, "bias": "center-left",   "category": "fact-checker"},
    "foxnews.com":          {"score": 0.45, "bias": "right",         "category": "mixed"},
    "dailymail.co.uk":      {"score": 0.40, "bias": "right",         "category": "mixed"},
    "breitbart.com":        {"score": 0.10, "bias": "extreme-right", "category": "fake"},
    "infowars.com":         {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "naturalnews.com":      {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "worldnewsdailyreport.com": {"score": 0.02, "bias": "fake",     "category": "fake"},
}
DEFAULT_INFO = {"score": 0.5, "bias": "unknown", "category": "unknown"}

SENSATIONAL_WORDS = [
    "shocking", "unbelievable", "breaking", "exclusive", "bombshell",
    "outrage", "scandal", "exposed", "revealed", "secret", "conspiracy",
    "urgent", "alert", "warning", "crisis", "disaster", "chaos",
    "incredible", "stunning", "explosive", "miracle", "banned",
    "censored", "truth", "wake up", "they don't want you to know"
]

def extract_domain(url_or_source):
    if not isinstance(url_or_source, str):
        return "unknown"
    try:
        url_or_source = url_or_source.strip()
        if url_or_source.startswith("http"):
            domain = urlparse(url_or_source).netloc.replace("www.", "")
            return domain if domain else "unknown"
        if "." in url_or_source:
            return url_or_source.replace("www.", "")
        return url_or_source
    except:
        return "unknown"

def get_mbfc_score(domain):
    domain = domain.lower().strip()
    if domain in MBFC_CREDIBILITY_DB:
        return MBFC_CREDIBILITY_DB[domain]
    for known, info in MBFC_CREDIBILITY_DB.items():
        if known in domain or domain in known:
            return info
    return DEFAULT_INFO

def compute_source_vector(domain_or_url, title, text=""):
    """
    Titre seul  → text=""  → vecteur [4]
    Titre+texte → text=... → vecteur [5]  (ajoute external_links)
    Choisir SOURCE_DIM en conséquence.
    """
    domain     = extract_domain(domain_or_url)
    mbfc_score = get_mbfc_score(domain)["score"]
    full       = title + " " + text
    nb_upper   = sum(1 for c in title if c.isupper())
    nb_letters = sum(1 for c in title if c.isalpha())
    ratio_up   = nb_upper / nb_letters if nb_letters > 0 else 0.0
    nb_excl    = full.count("!") + full.count("?")
    nb_words   = len(full.split())
    excl_sc    = nb_excl / nb_words if nb_words > 0 else 0.0
    nb_sens    = sum(1 for w in SENSATIONAL_WORDS if w in full.lower())
    sens_sc    = nb_sens / len(SENSATIONAL_WORDS)
    vec = [mbfc_score, ratio_up, excl_sc, sens_sc]
    if text:                                          # version titre+texte
        nb_links = len(re.findall(r'http[s]?://\S+', text))
        vec.append(min(nb_links / 10, 1.0))
    return np.array(vec, dtype=np.float32)

print("✅ Module 2 chargé")


# ============================================================
# CELLULE 4 — NER avec spaCy
# ============================================================

def extract_entities(title, text=""):
    """Extrait les entités nommées du titre (+ texte tronqué si fourni)."""
    text_part  = ' '.join(str(text).split()[:150]) if text else ""
    full_input = f"{title} {text_part}".strip()
    doc        = nlp(full_input)
    types_ok   = {"PERSON", "ORG", "GPE", "LOC", "EVENT"}
    seen, result = set(), []
    for ent in doc.ents:
        if ent.label_ in types_ok:
            t = ent.text.strip().strip("'s").strip()
            if len(t) > 1 and t.lower() not in seen:
                seen.add(t.lower())
                result.append((t, ent.label_))
    return result

print("✅ NER prêt")


# ============================================================
# CELLULE 5 — WIKIDATA : entity linking + récupération triplets
# ============================================================
# C'est ici que commence la vraie différence avec la version précédente.
# On ne se contente plus d'un hash du QID :
#   1. On trouve le QID de chaque entité (entity linking)
#   2. On récupère ses propriétés principales via l'API Wikidata
#      sous forme de triplets (sujet, relation, objet)
#   Ces triplets serviront à entraîner TransE.
# ============================================================

CACHE_FILE = "wikidata_cache_kg.pkl"

def load_cache():
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, 'rb') as f:
            return pickle.load(f)
    return {}

def save_cache(cache):
    with open(CACHE_FILE, 'wb') as f:
        pickle.dump(cache, f)

wikidata_cache = load_cache()
print(f"Cache chargé : {len(wikidata_cache)} entrées")

# ---- Entity Linking ----------------------------------------
def search_wikidata_entity(entity_text):
    """Renvoie le QID Wikidata de l'entité (ex: 'Q35'=Denmark)."""
    key = f"qid_{entity_text.lower().strip()}"
    if key in wikidata_cache:
        return wikidata_cache[key]
    try:
        r = requests.get(
            "https://www.wikidata.org/w/api.php",
            params={"action": "wbsearchentities", "search": entity_text,
                    "language": "en", "limit": 1, "format": "json"},
            timeout=5
        )
        data = r.json()
        qid  = data["search"][0]["id"] if data.get("search") else None
        wikidata_cache[key] = qid
        time.sleep(0.1)
        return qid
    except:
        wikidata_cache[key] = None
        return None

# ---- Récupération des triplets d'une entité ----------------
# On limite aux propriétés les plus informatives et fréquentes
# pour garder un graphe gérable (pas des millions de triplets).
IMPORTANT_PROPS = [
    "P31",   # instance of
    "P279",  # subclass of
    "P106",  # occupation
    "P27",   # country of citizenship
    "P17",   # country
    "P131",  # located in
    "P21",   # sex/gender
    "P19",   # place of birth
    "P495",  # country of origin
    "P136",  # genre
    "P571",  # inception
]

def get_entity_triplets(qid):
    """
    Récupère les triplets (qid, prop, value_qid) depuis Wikidata.
    Ne garde que les valeurs qui sont elles-mêmes des entités (QID).
    Renvoie une liste de tuples (sujet_str, relation_str, objet_str).
    """
    key = f"triplets_{qid}"
    if key in wikidata_cache:
        return wikidata_cache[key]
    if qid is None:
        return []
    try:
        r = requests.get(
            "https://www.wikidata.org/w/api.php",
            params={"action": "wbgetentities", "ids": qid,
                    "props": "claims", "format": "json"},
            timeout=8
        )
        data   = r.json()
        entity = data.get("entities", {}).get(qid, {})
        claims = entity.get("claims", {})
        triplets = []
        for prop in IMPORTANT_PROPS:
            if prop not in claims:
                continue
            for claim in claims[prop]:
                try:
                    snak = claim["mainsnak"]
                    if snak["snaktype"] == "value" and \
                       snak["datavalue"]["type"] == "wikibase-entityid":
                        obj_qid = snak["datavalue"]["value"]["id"]
                        triplets.append((qid, prop, obj_qid))
                except:
                    continue
        wikidata_cache[key] = triplets
        time.sleep(0.1)
        return triplets
    except:
        wikidata_cache[key] = []
        return []

print("✅ Fonctions Wikidata prêtes")


# ============================================================
# CELLULE 6 — CONSTRUCTION DU SOUS-GRAPHE + ENTRAÎNEMENT TransE
# ============================================================
# Idée : collecter tous les triplets des entités présentes dans
# le dataset → construire un graphe local → entraîner TransE
# avec PyKEEN → obtenir des embeddings "vrais" pour chaque entité.
#
# On fait ça UNE SEULE FOIS et on sauvegarde les embeddings.
# ============================================================

from pykeen.pipeline import pipeline as pykeen_pipeline
from pykeen.triples import TriplesFactory

def build_subgraph_and_train_transe(
        all_qids,
        embedding_dim   = 64,
        num_epochs      = 30,
        cache_path_emb  = "transe_embeddings.pkl",
        cache_path_trips= "subgraph_triplets.pkl"
):
    """
    1. Récupère les triplets Wikidata pour tous les QIDs du dataset.
    2. Construit un TriplesFactory PyKEEN.
    3. Entraîne TransE.
    4. Retourne un dict {qid: embedding_vector}.

    Si les embeddings sont déjà en cache, les charge directement.
    """
    # --- Chargement si déjà entraîné ---
    if os.path.exists(cache_path_emb):
        print(f"✅ Embeddings TransE déjà calculés — chargement depuis '{cache_path_emb}'")
        with open(cache_path_emb, 'rb') as f:
            return pickle.load(f)

    # --- Collecte des triplets ---
    if os.path.exists(cache_path_trips):
        print("✅ Triplets déjà collectés — chargement...")
        with open(cache_path_trips, 'rb') as f:
            all_triplets = pickle.load(f)
    else:
        print(f"Collecte des triplets Wikidata pour {len(all_qids)} entités...")
        all_triplets = []
        for i, qid in enumerate(all_qids):
            triplets = get_entity_triplets(qid)
            all_triplets.extend(triplets)
            if (i + 1) % 50 == 0:
                print(f"  {i+1}/{len(all_qids)} entités — {len(all_triplets)} triplets")
                save_cache(wikidata_cache)
        save_cache(wikidata_cache)
        with open(cache_path_trips, 'wb') as f:
            pickle.dump(all_triplets, f)
        print(f"✅ {len(all_triplets)} triplets collectés")

    if len(all_triplets) < 10:
        print("⚠️  Pas assez de triplets pour entraîner TransE — fallback zéros")
        return {}

    # --- Dédoublonnage ---
    all_triplets = list(set(all_triplets))
    print(f"   Triplets uniques : {len(all_triplets)}")

    # --- Conversion en tableau numpy [N, 3] de strings ---
    triples_array = np.array(all_triplets, dtype=str)  # (N, 3)

    # --- TriplesFactory PyKEEN ---
    tf = TriplesFactory.from_labeled_triples(triples_array)
    print(f"   Entités : {tf.num_entities} | Relations : {tf.num_relations}")

    # Séparation train/test (80/20) pour éviter data leakage dans PyKEEN
    train_tf, test_tf = tf.split([0.8, 0.2], random_state=42)

    # --- Entraînement TransE ---
    print(f"Entraînement TransE (dim={embedding_dim}, epochs={num_epochs})...")
    result = pykeen_pipeline(
        training   = train_tf,
        testing    = test_tf,
        model      = 'TransE',
        model_kwargs = dict(embedding_dim=embedding_dim),
        training_kwargs = dict(num_epochs=num_epochs, batch_size=512),
        optimizer  = 'Adam',
        optimizer_kwargs = dict(lr=0.01),
        random_seed = 42,
        device     = 'cuda' if torch.cuda.is_available() else 'cpu',
    )
    print("✅ TransE entraîné")
    print(f"   MR  : {result.metric_results.to_flat_dict().get('both.realistic.mean_rank', 'N/A')}")
    print(f"   H@10: {result.metric_results.to_flat_dict().get('both.realistic.hits_at_10', 'N/A')}")

    # --- Extraction des embeddings ---
    model_pykeen  = result.model
    entity_repr   = model_pykeen.entity_representations[0]
    # Récupère toutes les embeddings comme tenseur (num_entities, dim)
    all_emb       = entity_repr().detach().cpu().numpy()
    entity_to_id  = tf.entity_to_id          # dict {qid_str: int_index}

    emb_dict = {}
    for qid_str, idx in entity_to_id.items():
        emb_dict[qid_str] = all_emb[idx]

    # --- Sauvegarde ---
    with open(cache_path_emb, 'wb') as f:
        pickle.dump(emb_dict, f)
    print(f"✅ Embeddings sauvegardés dans '{cache_path_emb}'")
    return emb_dict


# ============================================================
# CELLULE 7 — VECTEUR KG PAR ARTICLE
# ============================================================
# Pour chaque article :
#   1. NER → liste d'entités
#   2. Entity linking → QIDs
#   3. Récupération de l'embedding TransE de chaque QID
#   4. Moyenne des embeddings → vecteur représentatif de l'article
#   + 4 features de couverture (ratio d'alignement, etc.)
# ============================================================

EMBEDDING_DIM = 64    # doit correspondre à build_subgraph_and_train_transe
KG_DIM        = EMBEDDING_DIM + 4   # 64 + 4 = 68

def get_kg_vector_for_article(title, text, transe_emb_dict):
    """
    Calcule le vecteur KG d'un article en utilisant les embeddings TransE.

    transe_emb_dict : dict {qid_str: np.array(embedding_dim)}
    """
    entities = extract_entities(title, text)
    nb_found = len(entities)

    if nb_found == 0:
        return np.zeros(KG_DIM, dtype=np.float32)

    qids = [search_wikidata_entity(ent) for ent, _ in entities]
    nb_aligned  = sum(1 for q in qids if q is not None)
    ratio_align = nb_aligned / nb_found if nb_found > 0 else 0.0

    embeddings = []
    for qid in qids:
        if qid and qid in transe_emb_dict:
            embeddings.append(transe_emb_dict[qid])
        elif qid:
            # QID connu mais pas dans le graphe entraîné → zéros
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))
        else:
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))

    mean_emb = np.mean(embeddings, axis=0).astype(np.float32)
    # Normalisation L2
    norm = np.linalg.norm(mean_emb)
    if norm > 0:
        mean_emb = mean_emb / norm

    coverage = np.array([
        min(nb_found    / 5.0, 1.0),   # densité d'entités
        ratio_align,                     # taux d'alignement Wikidata
        float(nb_aligned > 0),           # au moins une entité alignée
        min(nb_aligned  / 3.0, 1.0),    # couverture normalisée
    ], dtype=np.float32)

    return np.concatenate([mean_emb, coverage])

print(f"✅ KG_DIM = {KG_DIM}  (TransE {EMBEDDING_DIM}d + 4 coverage features)")


# ============================================================
# CELLULE 8 — PIPELINE COMPLÈTE : COLLECTE → GRAPH → EMBEDDINGS
# ============================================================

drive.mount('/content/drive')

DATASET_PATH      = '/content/drive/MyDrive/fakenewsnet_gossipcop_balanced.csv'
KG_CACHE_PATH     = '/content/drive/MyDrive/kg_vectors_transe_gossipcop.npy'
TRANSE_EMB_PATH   = '/content/drive/MyDrive/transe_embeddings_gossipcop.pkl'
TRIPLETS_PATH     = '/content/drive/MyDrive/subgraph_triplets_gossipcop.pkl'

# ---- Chargement dataset ----
df = pd.read_csv(DATASET_PATH)
if 'domain' not in df.columns:
    df['domain'] = df['url'].apply(
        lambda u: urlparse(str(u)).netloc.replace('www.', '') or 'unknown')
df['title']  = df['title'].fillna('')
df['text']   = df['text'].fillna('') if 'text' in df.columns else ''
df['domain'] = df['domain'].fillna('unknown').replace('nan', 'unknown')
df['label']  = df['label'].astype(int)
df = df.dropna(subset=['label']).reset_index(drop=True)

titles  = df['title'].tolist()
texts   = df['text'].apply(lambda t: ' '.join(str(t).split()[:150])).tolist() \
          if 'text' in df.columns else [''] * len(df)
labels  = df['label'].tolist()
domains = df['domain'].tolist()
print(f"Dataset : {len(titles)} articles")

# ---- Étape 1 : Entity linking sur tout le dataset ----
# On collecte tous les QIDs pour construire le sous-graphe.
print("\n[1/3] Entity linking sur tout le dataset...")
ALL_QIDS_CACHE = '/content/drive/MyDrive/all_qids_gossipcop.pkl'

if os.path.exists(ALL_QIDS_CACHE):
    with open(ALL_QIDS_CACHE, 'rb') as f:
        all_qids_set = pickle.load(f)
    print(f"✅ {len(all_qids_set)} QIDs chargés depuis le cache")
else:
    all_qids_set = set()
    for i, (title, text) in enumerate(zip(titles, texts)):
        entities = extract_entities(title, text)
        for ent, _ in entities:
            qid = search_wikidata_entity(ent)
            if qid:
                all_qids_set.add(qid)
        if (i + 1) % 200 == 0:
            print(f"  {i+1}/{len(titles)} articles — {len(all_qids_set)} QIDs uniques")
            save_cache(wikidata_cache)
    save_cache(wikidata_cache)
    with open(ALL_QIDS_CACHE, 'wb') as f:
        pickle.dump(all_qids_set, f)
    print(f"✅ {len(all_qids_set)} entités uniques trouvées")

# ---- Étape 2 : Construction du graphe + entraînement TransE ----
print("\n[2/3] Construction sous-graphe Wikidata + entraînement TransE...")
transe_emb_dict = build_subgraph_and_train_transe(
    all_qids        = list(all_qids_set),
    embedding_dim   = EMBEDDING_DIM,
    num_epochs      = 30,
    cache_path_emb  = TRANSE_EMB_PATH,
    cache_path_trips= TRIPLETS_PATH,
)
print(f"   Embeddings disponibles pour {len(transe_emb_dict)} entités")

# ---- Étape 3 : Précalcul des vecteurs KG pour chaque article ----
print("\n[3/3] Précalcul des vecteurs KG par article...")
if os.path.exists(KG_CACHE_PATH):
    kg_matrix = np.load(KG_CACHE_PATH)
    print(f"✅ KG vectors chargés : {kg_matrix.shape}")
else:
    kg_vectors = []
    for i, (title, text) in enumerate(zip(titles, texts)):
        kg_vectors.append(get_kg_vector_for_article(title, text, transe_emb_dict))
        if (i + 1) % 200 == 0:
            print(f"  {i+1}/{len(titles)}...")
            save_cache(wikidata_cache)
    kg_matrix = np.vstack(kg_vectors)
    np.save(KG_CACHE_PATH, kg_matrix)
    save_cache(wikidata_cache)
    print(f"✅ KG vectors calculés : {kg_matrix.shape}")

# ---- Vecteurs sources (Module 2) ----
# Choisir SOURCE_DIM=4 (titre seul) ou SOURCE_DIM=5 (titre+texte)
SOURCE_DIM = 4  # ← changer en 5 si version titre+texte
print("Calcul vecteurs sources (Module 2)...")
source_vectors = np.vstack([
    compute_source_vector(d, t, x if SOURCE_DIM == 5 else "")
    for d, t, x in zip(domains, titles, texts)
])
print(f"✅ Source vectors : {source_vectors.shape}")  # (N, 4) ou (N, 5)

print(f"\n{'='*50}")
print(f"RÉCAPITULATIF")
print(f"{'='*50}")
print(f"Articles       : {len(titles)}")
print(f"Source vectors : {source_vectors.shape}  ← Module 2")
print(f"KG vectors     : {kg_matrix.shape}  ← Module 3 (TransE réel)")
print(f"KG_DIM         : {KG_DIM}  ({EMBEDDING_DIM}d TransE + 4 coverage)")


# ============================================================
# CELLULE 9 — DATASET PYTORCH TRIMODAL
# ============================================================

class TrimodalDataset(Dataset):
    def __init__(self, texts, source_vectors, kg_vectors,
                 labels, tokenizer, max_length=128):
        self.texts          = texts
        self.source_vectors = source_vectors
        self.kg_vectors     = kg_vectors
        self.labels         = labels
        self.tokenizer      = tokenizer
        self.max_length     = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            str(self.texts[idx]),
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids':      enc['input_ids'].squeeze(),
            'attention_mask': enc['attention_mask'].squeeze(),
            'source_vector':  torch.tensor(self.source_vectors[idx], dtype=torch.float32),
            'kg_vector':      torch.tensor(self.kg_vectors[idx],     dtype=torch.float32),
            'label':          torch.tensor(self.labels[idx],         dtype=torch.long)
        }


# ============================================================
# CELLULE 10 — MODÈLE TRIMODAL
# Fusion BERT [768] + Source [32] + KG [64] = [864]
# ============================================================

class TrimodalFakeNewsDetector(nn.Module):
    def __init__(self, bert_model_name='bert-base-uncased',
                 source_dim=4,    # 4 (titre seul) ou 5 (titre+texte)
                 kg_dim=68,       # EMBEDDING_DIM + 4
                 hidden_dim=256,
                 num_classes=2,
                 dropout=0.3):
        super().__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        bert_hidden = self.bert.config.hidden_size   # 768

        self.source_encoder = nn.Sequential(
            nn.Linear(source_dim, 32), nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(32, 32), nn.ReLU()
        )
        self.kg_encoder = nn.Sequential(
            nn.Linear(kg_dim, 64), nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, 64), nn.ReLU()
        )
        # 768 + 32 + 64 = 864
        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + 32 + 64, hidden_dim),
            nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(), nn.Dropout(dropout / 2),
            nn.Linear(64, num_classes)
        )

    def forward(self, input_ids, attention_mask, source_vector, kg_vector):
        bert_out = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls  = bert_out.last_hidden_state[:, 0, :]
        src  = self.source_encoder(source_vector)
        kg   = self.kg_encoder(kg_vector)
        fused = torch.cat([cls, src, kg], dim=1)
        return self.classifier(fused)


# ============================================================
# CELLULE 11 — ENTRAÎNEMENT ET ÉVALUATION
# ============================================================

def train_epoch(model, loader, optimizer, scheduler, criterion, device):
    model.train()
    total_loss, all_preds, all_labels_list = 0, [], []
    for batch in loader:
        ids   = batch['input_ids'].to(device)
        mask  = batch['attention_mask'].to(device)
        src   = batch['source_vector'].to(device)
        kg    = batch['kg_vector'].to(device)
        lbls  = batch['label'].to(device)
        optimizer.zero_grad()
        logits = model(ids, mask, src, kg)
        loss   = criterion(logits, lbls)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()
        all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
        all_labels_list.extend(lbls.cpu().numpy())
    return total_loss / len(loader), accuracy_score(all_labels_list, all_preds)

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, all_preds, all_labels_list, all_probs = 0, [], [], []
    with torch.no_grad():
        for batch in loader:
            ids   = batch['input_ids'].to(device)
            mask  = batch['attention_mask'].to(device)
            src   = batch['source_vector'].to(device)
            kg    = batch['kg_vector'].to(device)
            lbls  = batch['label'].to(device)
            logits = model(ids, mask, src, kg)
            loss   = criterion(logits, lbls)
            total_loss += loss.item()
            all_probs.extend(torch.softmax(logits, 1)[:, 1].cpu().numpy())
            all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
            all_labels_list.extend(lbls.cpu().numpy())
    acc = accuracy_score(all_labels_list, all_preds)
    f1  = f1_score(all_labels_list, all_preds, average='weighted')
    auc = roc_auc_score(all_labels_list, all_probs)
    return total_loss / len(loader), acc, f1, auc, all_preds, all_labels_list


# ============================================================
# CELLULE 12 — PIPELINE PRINCIPALE
# ============================================================

MAX_LENGTH    = 128    # titre seul → 128 | titre+texte → 256
BATCH_SIZE    = 16
EPOCHS        = 3
LEARNING_RATE = 2e-5

# Pour version titre seul : utilise titles comme input BERT
# Pour version titre+texte : utilise contents = [f"{t} {x}" for t,x in zip(titles,texts)]
bert_inputs = titles   # ← remplacer par contents pour titre+texte

(train_texts, test_texts,
 train_src,   test_src,
 train_kg,    test_kg,
 train_labels, test_labels) = train_test_split(
    bert_inputs, source_vectors, kg_matrix, labels,
    test_size=0.2, random_state=42, stratify=labels
)
print(f"Train : {len(train_texts)} | Test : {len(test_texts)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

train_ds = TrimodalDataset(train_texts, train_src, train_kg, train_labels, tokenizer, MAX_LENGTH)
test_ds  = TrimodalDataset(test_texts,  test_src,  test_kg,  test_labels,  tokenizer, MAX_LENGTH)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False)

model = TrimodalFakeNewsDetector(
    source_dim = SOURCE_DIM,   # 4 ou 5
    kg_dim     = KG_DIM        # EMBEDDING_DIM + 4 = 68
).to(device)
print(f"Paramètres totaux : {sum(p.numel() for p in model.parameters()):,}")

class_weights = compute_class_weight('balanced',
                                      classes=np.array([0, 1]),
                                      y=train_labels)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor(class_weights, dtype=torch.float32).to(device))
optimizer   = AdamW(model.parameters(), lr=LEARNING_RATE)
total_steps = len(train_loader) * EPOCHS
scheduler   = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps   = total_steps // 10,
    num_training_steps = total_steps
)

print("\n=== ENTRAÎNEMENT M1+M2+M3 (TransE réels) ===\n")
history  = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
best_f1  = 0

for epoch in range(EPOCHS):
    print(f"--- Époque {epoch+1}/{EPOCHS} ---")
    tr_loss, tr_acc = train_epoch(model, train_loader, optimizer, scheduler, criterion, device)
    vl_loss, vl_acc, vl_f1, vl_auc, _, _ = evaluate(model, test_loader, criterion, device)
    history['train_loss'].append(tr_loss)
    history['train_acc'].append(tr_acc)
    history['val_loss'].append(vl_loss)
    history['val_acc'].append(vl_acc)
    print(f"Train → Loss: {tr_loss:.4f} | Acc: {tr_acc:.4f}")
    print(f"Test  → Loss: {vl_loss:.4f} | Acc: {vl_acc:.4f} | F1: {vl_f1:.4f} | AUC: {vl_auc:.4f}")
    if vl_f1 > best_f1:
        best_f1 = vl_f1
        torch.save(model.state_dict(), 'best_trimodal_transe.pt')
        print(f"  ✅ Meilleur modèle (F1={best_f1:.4f})\n")
    else:
        print()

model.load_state_dict(torch.load('best_trimodal_transe.pt'))
print("\n=== ÉVALUATION FINALE ===\n")
_, acc, f1, auc, preds, true_labels = evaluate(model, test_loader, criterion, device)
print(classification_report(true_labels, preds, target_names=['Réel', 'Fake']))
print(f"Accuracy : {acc:.4f} | F1 : {f1:.4f} | AUC : {auc:.4f}")

# --- Courbes ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(history['train_loss'], label='Train', marker='o')
ax1.plot(history['val_loss'],   label='Val',   marker='o')
ax1.set_title('Loss — M1+M2+M3 (TransE)') ; ax1.legend()
ax2.plot(history['train_acc'], label='Train', marker='o')
ax2.plot(history['val_acc'],   label='Val',   marker='o')
ax2.set_title('Accuracy — M1+M2+M3 (TransE)') ; ax2.legend()
plt.tight_layout()
plt.savefig('trimodal_transe_curves.png', dpi=120)
plt.show()

# --- Tableau comparatif ---
print("\n" + "="*65)
print("TABLEAU COMPARATIF — GossipCop")
print("="*65)
print(f"{'Modèle':<42} {'Accuracy':>10} {'F1':>8} {'AUC':>8}")
print("-"*65)
print(f"{'TF-IDF+SVM (titre seul)':<42} {'0.7876':>10} {'0.7875':>8} {'0.8599':>8}")
print(f"{'TF-IDF+SVM (titre+texte)':<42} {'0.8293':>10} {'0.8293':>8} {'0.9066':>8}")
print(f"{'BERT+Source M1+M2 (titre seul)':<42} {'0.8235':>10} {'0.8234':>8} {'0.8942':>8}")
print(f"{'M1+M2+M3 hash (titre seul)':<42} {'⏳':>10} {'⏳':>8} {'⏳':>8}")
print(f"{'M1+M2+M3 TransE (titre seul)':<42} {acc:>10.4f} {f1:>8.4f} {auc:>8.4f}")
print("="*65)

torch.save(model.state_dict(), 'trimodal_transe_final.pt')
print("✅ Modèle final sauvegardé")
```

```python
# ============================================================
# MODULE 3 — VRAIS GRAPH EMBEDDINGS (TransE via PyKEEN)
# VERSION : TITRE + TEXTE
# Approche : sous-graphe Wikidata local → entraînement TransE
# source_dim=5 | MAX_LENGTH=256 | bert_inputs=contents
# ============================================================

# ============================================================
# CELLULE 1 — INSTALLATION
# ============================================================
!pip install spacy pykeen requests -q
!python -m spacy download en_core_web_sm -q
print("✅ Installations terminées")


# ============================================================
# CELLULE 2 — IMPORTS
# ============================================================
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel
from transformers import get_linear_schedule_with_warmup
from torch.optim import AdamW
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, f1_score,
                             roc_auc_score, classification_report)
from sklearn.utils.class_weight import compute_class_weight
import spacy, requests, time, os, pickle, re
from urllib.parse import urlparse
from google.colab import drive
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Appareil utilisé : {device}")
nlp = spacy.load("en_core_web_sm")
print("✅ spaCy chargé")


# ============================================================
# CELLULE 3 — MODULE 2 INTÉGRÉ (titre seul, dim=4)
# Remplace source_dim=5 et les deux dernières lignes si tu veux
# la version titre+texte — le reste du code ne change pas.
# ============================================================

MBFC_CREDIBILITY_DB = {
    "reuters.com":          {"score": 0.95, "bias": "center",        "category": "reliable"},
    "apnews.com":           {"score": 0.95, "bias": "center",        "category": "reliable"},
    "bbc.com":              {"score": 0.90, "bias": "center-left",   "category": "reliable"},
    "theguardian.com":      {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "nytimes.com":          {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "washingtonpost.com":   {"score": 0.85, "bias": "center-left",   "category": "reliable"},
    "npr.org":              {"score": 0.88, "bias": "center-left",   "category": "reliable"},
    "politifact.com":       {"score": 0.92, "bias": "center",        "category": "fact-checker"},
    "snopes.com":           {"score": 0.90, "bias": "center-left",   "category": "fact-checker"},
    "foxnews.com":          {"score": 0.45, "bias": "right",         "category": "mixed"},
    "dailymail.co.uk":      {"score": 0.40, "bias": "right",         "category": "mixed"},
    "breitbart.com":        {"score": 0.10, "bias": "extreme-right", "category": "fake"},
    "infowars.com":         {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "naturalnews.com":      {"score": 0.05, "bias": "extreme-right", "category": "fake"},
    "worldnewsdailyreport.com": {"score": 0.02, "bias": "fake",     "category": "fake"},
}
DEFAULT_INFO = {"score": 0.5, "bias": "unknown", "category": "unknown"}

SENSATIONAL_WORDS = [
    "shocking", "unbelievable", "breaking", "exclusive", "bombshell",
    "outrage", "scandal", "exposed", "revealed", "secret", "conspiracy",
    "urgent", "alert", "warning", "crisis", "disaster", "chaos",
    "incredible", "stunning", "explosive", "miracle", "banned",
    "censored", "truth", "wake up", "they don't want you to know"
]

def extract_domain(url_or_source):
    if not isinstance(url_or_source, str):
        return "unknown"
    try:
        url_or_source = url_or_source.strip()
        if url_or_source.startswith("http"):
            domain = urlparse(url_or_source).netloc.replace("www.", "")
            return domain if domain else "unknown"
        if "." in url_or_source:
            return url_or_source.replace("www.", "")
        return url_or_source
    except:
        return "unknown"

def get_mbfc_score(domain):
    domain = domain.lower().strip()
    if domain in MBFC_CREDIBILITY_DB:
        return MBFC_CREDIBILITY_DB[domain]
    for known, info in MBFC_CREDIBILITY_DB.items():
        if known in domain or domain in known:
            return info
    return DEFAULT_INFO

def compute_source_vector(domain_or_url, title, text=""):
    """
    Titre seul  → text=""  → vecteur [4]
    Titre+texte → text=... → vecteur [5]  (ajoute external_links)
    Choisir SOURCE_DIM en conséquence.
    """
    domain     = extract_domain(domain_or_url)
    mbfc_score = get_mbfc_score(domain)["score"]
    full       = title + " " + text
    nb_upper   = sum(1 for c in title if c.isupper())
    nb_letters = sum(1 for c in title if c.isalpha())
    ratio_up   = nb_upper / nb_letters if nb_letters > 0 else 0.0
    nb_excl    = full.count("!") + full.count("?")
    nb_words   = len(full.split())
    excl_sc    = nb_excl / nb_words if nb_words > 0 else 0.0
    nb_sens    = sum(1 for w in SENSATIONAL_WORDS if w in full.lower())
    sens_sc    = nb_sens / len(SENSATIONAL_WORDS)
    vec = [mbfc_score, ratio_up, excl_sc, sens_sc]
    if text:                                          # version titre+texte
        nb_links = len(re.findall(r'http[s]?://\S+', text))
        vec.append(min(nb_links / 10, 1.0))
    return np.array(vec, dtype=np.float32)

print("✅ Module 2 chargé")


# ============================================================
# CELLULE 4 — NER avec spaCy
# ============================================================

def extract_entities(title, text=""):
    """Extrait les entités nommées du titre (+ texte tronqué si fourni)."""
    text_part  = ' '.join(str(text).split()[:150]) if text else ""
    full_input = f"{title} {text_part}".strip()
    doc        = nlp(full_input)
    types_ok   = {"PERSON", "ORG", "GPE", "LOC", "EVENT"}
    seen, result = set(), []
    for ent in doc.ents:
        if ent.label_ in types_ok:
            t = ent.text.strip().strip("'s").strip()
            if len(t) > 1 and t.lower() not in seen:
                seen.add(t.lower())
                result.append((t, ent.label_))
    return result

print("✅ NER prêt")


# ============================================================
# CELLULE 5 — WIKIDATA : entity linking + récupération triplets
# ============================================================
# C'est ici que commence la vraie différence avec la version précédente.
# On ne se contente plus d'un hash du QID :
#   1. On trouve le QID de chaque entité (entity linking)
#   2. On récupère ses propriétés principales via l'API Wikidata
#      sous forme de triplets (sujet, relation, objet)
#   Ces triplets serviront à entraîner TransE.
# ============================================================

CACHE_FILE = "wikidata_cache_kg.pkl"

def load_cache():
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, 'rb') as f:
            return pickle.load(f)
    return {}

def save_cache(cache):
    with open(CACHE_FILE, 'wb') as f:
        pickle.dump(cache, f)

wikidata_cache = load_cache()
print(f"Cache chargé : {len(wikidata_cache)} entrées")

# ---- Entity Linking ----------------------------------------
def search_wikidata_entity(entity_text):
    """Renvoie le QID Wikidata de l'entité (ex: 'Q35'=Denmark)."""
    key = f"qid_{entity_text.lower().strip()}"
    if key in wikidata_cache:
        return wikidata_cache[key]
    try:
        r = requests.get(
            "https://www.wikidata.org/w/api.php",
            params={"action": "wbsearchentities", "search": entity_text,
                    "language": "en", "limit": 1, "format": "json"},
            timeout=5
        )
        data = r.json()
        qid  = data["search"][0]["id"] if data.get("search") else None
        wikidata_cache[key] = qid
        time.sleep(0.1)
        return qid
    except:
        wikidata_cache[key] = None
        return None

# ---- Récupération des triplets d'une entité ----------------
# On limite aux propriétés les plus informatives et fréquentes
# pour garder un graphe gérable (pas des millions de triplets).
IMPORTANT_PROPS = [
    "P31",   # instance of
    "P279",  # subclass of
    "P106",  # occupation
    "P27",   # country of citizenship
    "P17",   # country
    "P131",  # located in
    "P21",   # sex/gender
    "P19",   # place of birth
    "P495",  # country of origin
    "P136",  # genre
    "P571",  # inception
]

def get_entity_triplets(qid):
    """
    Récupère les triplets (qid, prop, value_qid) depuis Wikidata.
    Ne garde que les valeurs qui sont elles-mêmes des entités (QID).
    Renvoie une liste de tuples (sujet_str, relation_str, objet_str).
    """
    key = f"triplets_{qid}"
    if key in wikidata_cache:
        return wikidata_cache[key]
    if qid is None:
        return []
    try:
        r = requests.get(
            "https://www.wikidata.org/w/api.php",
            params={"action": "wbgetentities", "ids": qid,
                    "props": "claims", "format": "json"},
            timeout=8
        )
        data   = r.json()
        entity = data.get("entities", {}).get(qid, {})
        claims = entity.get("claims", {})
        triplets = []
        for prop in IMPORTANT_PROPS:
            if prop not in claims:
                continue
            for claim in claims[prop]:
                try:
                    snak = claim["mainsnak"]
                    if snak["snaktype"] == "value" and \
                       snak["datavalue"]["type"] == "wikibase-entityid":
                        obj_qid = snak["datavalue"]["value"]["id"]
                        triplets.append((qid, prop, obj_qid))
                except:
                    continue
        wikidata_cache[key] = triplets
        time.sleep(0.1)
        return triplets
    except:
        wikidata_cache[key] = []
        return []

print("✅ Fonctions Wikidata prêtes")


# ============================================================
# CELLULE 6 — CONSTRUCTION DU SOUS-GRAPHE + ENTRAÎNEMENT TransE
# ============================================================
# Idée : collecter tous les triplets des entités présentes dans
# le dataset → construire un graphe local → entraîner TransE
# avec PyKEEN → obtenir des embeddings "vrais" pour chaque entité.
#
# On fait ça UNE SEULE FOIS et on sauvegarde les embeddings.
# ============================================================

from pykeen.pipeline import pipeline as pykeen_pipeline
from pykeen.triples import TriplesFactory

def build_subgraph_and_train_transe(
        all_qids,
        embedding_dim   = 64,
        num_epochs      = 30,
        cache_path_emb  = "transe_embeddings.pkl",
        cache_path_trips= "subgraph_triplets.pkl"
):
    """
    1. Récupère les triplets Wikidata pour tous les QIDs du dataset.
    2. Construit un TriplesFactory PyKEEN.
    3. Entraîne TransE.
    4. Retourne un dict {qid: embedding_vector}.

    Si les embeddings sont déjà en cache, les charge directement.
    """
    # --- Chargement si déjà entraîné ---
    if os.path.exists(cache_path_emb):
        print(f"✅ Embeddings TransE déjà calculés — chargement depuis '{cache_path_emb}'")
        with open(cache_path_emb, 'rb') as f:
            return pickle.load(f)

    # --- Collecte des triplets ---
    if os.path.exists(cache_path_trips):
        print("✅ Triplets déjà collectés — chargement...")
        with open(cache_path_trips, 'rb') as f:
            all_triplets = pickle.load(f)
    else:
        print(f"Collecte des triplets Wikidata pour {len(all_qids)} entités...")
        all_triplets = []
        for i, qid in enumerate(all_qids):
            triplets = get_entity_triplets(qid)
            all_triplets.extend(triplets)
            if (i + 1) % 50 == 0:
                print(f"  {i+1}/{len(all_qids)} entités — {len(all_triplets)} triplets")
                save_cache(wikidata_cache)
        save_cache(wikidata_cache)
        with open(cache_path_trips, 'wb') as f:
            pickle.dump(all_triplets, f)
        print(f"✅ {len(all_triplets)} triplets collectés")

    if len(all_triplets) < 10:
        print("⚠️  Pas assez de triplets pour entraîner TransE — fallback zéros")
        return {}

    # --- Dédoublonnage ---
    all_triplets = list(set(all_triplets))
    print(f"   Triplets uniques : {len(all_triplets)}")

    # --- Conversion en tableau numpy [N, 3] de strings ---
    triples_array = np.array(all_triplets, dtype=str)  # (N, 3)

    # --- TriplesFactory PyKEEN ---
    tf = TriplesFactory.from_labeled_triples(triples_array)
    print(f"   Entités : {tf.num_entities} | Relations : {tf.num_relations}")

    # Séparation train/test (80/20) pour éviter data leakage dans PyKEEN
    train_tf, test_tf = tf.split([0.8, 0.2], random_state=42)

    # --- Entraînement TransE ---
    print(f"Entraînement TransE (dim={embedding_dim}, epochs={num_epochs})...")
    result = pykeen_pipeline(
        training   = train_tf,
        testing    = test_tf,
        model      = 'TransE',
        model_kwargs = dict(embedding_dim=embedding_dim),
        training_kwargs = dict(num_epochs=num_epochs, batch_size=512),
        optimizer  = 'Adam',
        optimizer_kwargs = dict(lr=0.01),
        random_seed = 42,
        device     = 'cuda' if torch.cuda.is_available() else 'cpu',
    )
    print("✅ TransE entraîné")
    print(f"   MR  : {result.metric_results.to_flat_dict().get('both.realistic.mean_rank', 'N/A')}")
    print(f"   H@10: {result.metric_results.to_flat_dict().get('both.realistic.hits_at_10', 'N/A')}")

    # --- Extraction des embeddings ---
    model_pykeen  = result.model
    entity_repr   = model_pykeen.entity_representations[0]
    # Récupère toutes les embeddings comme tenseur (num_entities, dim)
    all_emb       = entity_repr().detach().cpu().numpy()
    entity_to_id  = tf.entity_to_id          # dict {qid_str: int_index}

    emb_dict = {}
    for qid_str, idx in entity_to_id.items():
        emb_dict[qid_str] = all_emb[idx]

    # --- Sauvegarde ---
    with open(cache_path_emb, 'wb') as f:
        pickle.dump(emb_dict, f)
    print(f"✅ Embeddings sauvegardés dans '{cache_path_emb}'")
    return emb_dict


# ============================================================
# CELLULE 7 — VECTEUR KG PAR ARTICLE
# ============================================================
# Pour chaque article :
#   1. NER → liste d'entités
#   2. Entity linking → QIDs
#   3. Récupération de l'embedding TransE de chaque QID
#   4. Moyenne des embeddings → vecteur représentatif de l'article
#   + 4 features de couverture (ratio d'alignement, etc.)
# ============================================================

EMBEDDING_DIM = 64    # doit correspondre à build_subgraph_and_train_transe
KG_DIM        = EMBEDDING_DIM + 4   # 64 + 4 = 68

def get_kg_vector_for_article(title, text, transe_emb_dict):
    """
    Calcule le vecteur KG d'un article en utilisant les embeddings TransE.

    transe_emb_dict : dict {qid_str: np.array(embedding_dim)}
    """
    entities = extract_entities(title, text)
    nb_found = len(entities)

    if nb_found == 0:
        return np.zeros(KG_DIM, dtype=np.float32)

    qids = [search_wikidata_entity(ent) for ent, _ in entities]
    nb_aligned  = sum(1 for q in qids if q is not None)
    ratio_align = nb_aligned / nb_found if nb_found > 0 else 0.0

    embeddings = []
    for qid in qids:
        if qid and qid in transe_emb_dict:
            embeddings.append(transe_emb_dict[qid])
        elif qid:
            # QID connu mais pas dans le graphe entraîné → zéros
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))
        else:
            embeddings.append(np.zeros(EMBEDDING_DIM, dtype=np.float32))

    mean_emb = np.mean(embeddings, axis=0).astype(np.float32)
    # Normalisation L2
    norm = np.linalg.norm(mean_emb)
    if norm > 0:
        mean_emb = mean_emb / norm

    coverage = np.array([
        min(nb_found    / 5.0, 1.0),   # densité d'entités
        ratio_align,                     # taux d'alignement Wikidata
        float(nb_aligned > 0),           # au moins une entité alignée
        min(nb_aligned  / 3.0, 1.0),    # couverture normalisée
    ], dtype=np.float32)

    return np.concatenate([mean_emb, coverage])

print(f"✅ KG_DIM = {KG_DIM}  (TransE {EMBEDDING_DIM}d + 4 coverage features)")


# ============================================================
# CELLULE 8 — PIPELINE COMPLÈTE : COLLECTE → GRAPH → EMBEDDINGS
# ============================================================

drive.mount('/content/drive')

DATASET_PATH      = '/content/drive/MyDrive/fakenewsnet_gossipcop_balanced.csv'
KG_CACHE_PATH     = '/content/drive/MyDrive/kg_vectors_transe_gossipcop_tt.npy'
TRANSE_EMB_PATH   = '/content/drive/MyDrive/transe_embeddings_gossipcop_tt.pkl'
TRIPLETS_PATH     = '/content/drive/MyDrive/subgraph_triplets_gossipcop_tt.pkl'

# ---- Chargement dataset ----
df = pd.read_csv(DATASET_PATH)
if 'domain' not in df.columns:
    df['domain'] = df['url'].apply(
        lambda u: urlparse(str(u)).netloc.replace('www.', '') or 'unknown')
df['title']  = df['title'].fillna('')
df['text']   = df['text'].fillna('') if 'text' in df.columns else ''
df['domain'] = df['domain'].fillna('unknown').replace('nan', 'unknown')
df['label']  = df['label'].astype(int)
df = df.dropna(subset=['label']).reset_index(drop=True)

titles  = df['title'].tolist()
texts   = df['text'].apply(lambda t: ' '.join(str(t).split()[:150])).tolist() \
          if 'text' in df.columns else [''] * len(df)
labels  = df['label'].tolist()
domains = df['domain'].tolist()
print(f"Dataset : {len(titles)} articles")

# ---- Étape 1 : Entity linking sur tout le dataset ----
# On collecte tous les QIDs pour construire le sous-graphe.
print("\n[1/3] Entity linking sur tout le dataset...")
ALL_QIDS_CACHE = '/content/drive/MyDrive/all_qids_gossipcop_tt.pkl'

if os.path.exists(ALL_QIDS_CACHE):
    with open(ALL_QIDS_CACHE, 'rb') as f:
        all_qids_set = pickle.load(f)
    print(f"✅ {len(all_qids_set)} QIDs chargés depuis le cache")
else:
    all_qids_set = set()
    for i, (title, text) in enumerate(zip(titles, texts)):
        entities = extract_entities(title, text)
        for ent, _ in entities:
            qid = search_wikidata_entity(ent)
            if qid:
                all_qids_set.add(qid)
        if (i + 1) % 200 == 0:
            print(f"  {i+1}/{len(titles)} articles — {len(all_qids_set)} QIDs uniques")
            save_cache(wikidata_cache)
    save_cache(wikidata_cache)
    with open(ALL_QIDS_CACHE, 'wb') as f:
        pickle.dump(all_qids_set, f)
    print(f"✅ {len(all_qids_set)} entités uniques trouvées")

# ---- Étape 2 : Construction du graphe + entraînement TransE ----
print("\n[2/3] Construction sous-graphe Wikidata + entraînement TransE...")
transe_emb_dict = build_subgraph_and_train_transe(
    all_qids        = list(all_qids_set),
    embedding_dim   = EMBEDDING_DIM,
    num_epochs      = 30,
    cache_path_emb  = TRANSE_EMB_PATH,
    cache_path_trips= TRIPLETS_PATH,
)
print(f"   Embeddings disponibles pour {len(transe_emb_dict)} entités")

# ---- Étape 3 : Précalcul des vecteurs KG pour chaque article ----
print("\n[3/3] Précalcul des vecteurs KG par article...")
if os.path.exists(KG_CACHE_PATH):
    kg_matrix = np.load(KG_CACHE_PATH)
    print(f"✅ KG vectors chargés : {kg_matrix.shape}")
else:
    kg_vectors = []
    for i, (title, text) in enumerate(zip(titles, texts)):
        kg_vectors.append(get_kg_vector_for_article(title, text, transe_emb_dict))
        if (i + 1) % 200 == 0:
            print(f"  {i+1}/{len(titles)}...")
            save_cache(wikidata_cache)
    kg_matrix = np.vstack(kg_vectors)
    np.save(KG_CACHE_PATH, kg_matrix)
    save_cache(wikidata_cache)
    print(f"✅ KG vectors calculés : {kg_matrix.shape}")

# ---- Vecteurs sources (Module 2) ----
# Choisir SOURCE_DIM=4 (titre seul) ou SOURCE_DIM=5 (titre+texte)
SOURCE_DIM = 5  # ✅ titre+texte
print("Calcul vecteurs sources (Module 2)...")
source_vectors = np.vstack([
    compute_source_vector(d, t, x)   # ✅ toujours titre+texte
    for d, t, x in zip(domains, titles, texts)
])
print(f"✅ Source vectors : {source_vectors.shape}")  # (N, 4) ou (N, 5)

print(f"\n{'='*50}")
print(f"RÉCAPITULATIF")
print(f"{'='*50}")
print(f"Articles       : {len(titles)}")
print(f"Source vectors : {source_vectors.shape}  ← Module 2")
print(f"KG vectors     : {kg_matrix.shape}  ← Module 3 (TransE réel)")
print(f"KG_DIM         : {KG_DIM}  ({EMBEDDING_DIM}d TransE + 4 coverage)")


# ============================================================
# CELLULE 9 — DATASET PYTORCH TRIMODAL
# ============================================================

class TrimodalDataset(Dataset):
    def __init__(self, texts, source_vectors, kg_vectors,
                 labels, tokenizer, max_length=128):
        self.texts          = texts
        self.source_vectors = source_vectors
        self.kg_vectors     = kg_vectors
        self.labels         = labels
        self.tokenizer      = tokenizer
        self.max_length     = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        enc = self.tokenizer(
            str(self.texts[idx]),
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids':      enc['input_ids'].squeeze(),
            'attention_mask': enc['attention_mask'].squeeze(),
            'source_vector':  torch.tensor(self.source_vectors[idx], dtype=torch.float32),
            'kg_vector':      torch.tensor(self.kg_vectors[idx],     dtype=torch.float32),
            'label':          torch.tensor(self.labels[idx],         dtype=torch.long)
        }


# ============================================================
# CELLULE 10 — MODÈLE TRIMODAL
# Fusion BERT [768] + Source [32] + KG [64] = [864]
# ============================================================

class TrimodalFakeNewsDetector(nn.Module):
    def __init__(self, bert_model_name='bert-base-uncased',
                 source_dim=5,    # ✅ titre+texte
                 kg_dim=68,       # EMBEDDING_DIM + 4
                 hidden_dim=256,
                 num_classes=2,
                 dropout=0.3):
        super().__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        bert_hidden = self.bert.config.hidden_size   # 768

        self.source_encoder = nn.Sequential(
            nn.Linear(source_dim, 32), nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(32, 32), nn.ReLU()
        )
        self.kg_encoder = nn.Sequential(
            nn.Linear(kg_dim, 64), nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(64, 64), nn.ReLU()
        )
        # 768 + 32 + 64 = 864
        self.classifier = nn.Sequential(
            nn.Linear(bert_hidden + 32 + 64, hidden_dim),
            nn.ReLU(), nn.Dropout(dropout),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(), nn.Dropout(dropout / 2),
            nn.Linear(64, num_classes)
        )

    def forward(self, input_ids, attention_mask, source_vector, kg_vector):
        bert_out = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls  = bert_out.last_hidden_state[:, 0, :]
        src  = self.source_encoder(source_vector)
        kg   = self.kg_encoder(kg_vector)
        fused = torch.cat([cls, src, kg], dim=1)
        return self.classifier(fused)


# ============================================================
# CELLULE 11 — ENTRAÎNEMENT ET ÉVALUATION
# ============================================================

def train_epoch(model, loader, optimizer, scheduler, criterion, device):
    model.train()
    total_loss, all_preds, all_labels_list = 0, [], []
    for batch in loader:
        ids   = batch['input_ids'].to(device)
        mask  = batch['attention_mask'].to(device)
        src   = batch['source_vector'].to(device)
        kg    = batch['kg_vector'].to(device)
        lbls  = batch['label'].to(device)
        optimizer.zero_grad()
        logits = model(ids, mask, src, kg)
        loss   = criterion(logits, lbls)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()
        all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
        all_labels_list.extend(lbls.cpu().numpy())
    return total_loss / len(loader), accuracy_score(all_labels_list, all_preds)

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, all_preds, all_labels_list, all_probs = 0, [], [], []
    with torch.no_grad():
        for batch in loader:
            ids   = batch['input_ids'].to(device)
            mask  = batch['attention_mask'].to(device)
            src   = batch['source_vector'].to(device)
            kg    = batch['kg_vector'].to(device)
            lbls  = batch['label'].to(device)
            logits = model(ids, mask, src, kg)
            loss   = criterion(logits, lbls)
            total_loss += loss.item()
            all_probs.extend(torch.softmax(logits, 1)[:, 1].cpu().numpy())
            all_preds.extend(torch.argmax(logits, 1).cpu().numpy())
            all_labels_list.extend(lbls.cpu().numpy())
    acc = accuracy_score(all_labels_list, all_preds)
    f1  = f1_score(all_labels_list, all_preds, average='weighted')
    auc = roc_auc_score(all_labels_list, all_probs)
    return total_loss / len(loader), acc, f1, auc, all_preds, all_labels_list


# ============================================================
# CELLULE 12 — PIPELINE PRINCIPALE
# ============================================================

MAX_LENGTH    = 256    # ✅ titre+texte
BATCH_SIZE    = 16
EPOCHS        = 3
LEARNING_RATE = 2e-5

# ✅ VERSION TITRE+TEXTE — BERT reçoit titre + 150 premiers mots du texte
contents    = [f"{t} {x}".strip() for t, x in zip(titles, texts)]
bert_inputs = contents

(train_texts, test_texts,
 train_src,   test_src,
 train_kg,    test_kg,
 train_labels, test_labels) = train_test_split(
    bert_inputs, source_vectors, kg_matrix, labels,
    test_size=0.2, random_state=42, stratify=labels
)
print(f"Train : {len(train_texts)} | Test : {len(test_texts)}")

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

train_ds = TrimodalDataset(train_texts, train_src, train_kg, train_labels, tokenizer, MAX_LENGTH)
test_ds  = TrimodalDataset(test_texts,  test_src,  test_kg,  test_labels,  tokenizer, MAX_LENGTH)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False)

model = TrimodalFakeNewsDetector(
    source_dim = SOURCE_DIM,   # ✅ 5
    kg_dim     = KG_DIM        # EMBEDDING_DIM + 4 = 68
).to(device)
print(f"Paramètres totaux : {sum(p.numel() for p in model.parameters()):,}")

class_weights = compute_class_weight('balanced',
                                      classes=np.array([0, 1]),
                                      y=train_labels)
criterion = nn.CrossEntropyLoss(
    weight=torch.tensor(class_weights, dtype=torch.float32).to(device))
optimizer   = AdamW(model.parameters(), lr=LEARNING_RATE)
total_steps = len(train_loader) * EPOCHS
scheduler   = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps   = total_steps // 10,
    num_training_steps = total_steps
)

print("\n=== ENTRAÎNEMENT M1+M2+M3 (TransE — Titre+Texte) ===\n")
history  = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
best_f1  = 0

for epoch in range(EPOCHS):
    print(f"--- Époque {epoch+1}/{EPOCHS} ---")
    tr_loss, tr_acc = train_epoch(model, train_loader, optimizer, scheduler, criterion, device)
    vl_loss, vl_acc, vl_f1, vl_auc, _, _ = evaluate(model, test_loader, criterion, device)
    history['train_loss'].append(tr_loss)
    history['train_acc'].append(tr_acc)
    history['val_loss'].append(vl_loss)
    history['val_acc'].append(vl_acc)
    print(f"Train → Loss: {tr_loss:.4f} | Acc: {tr_acc:.4f}")
    print(f"Test  → Loss: {vl_loss:.4f} | Acc: {vl_acc:.4f} | F1: {vl_f1:.4f} | AUC: {vl_auc:.4f}")
    if vl_f1 > best_f1:
        best_f1 = vl_f1
        torch.save(model.state_dict(), 'best_trimodal_transe_tt.pt')
        print(f"  ✅ Meilleur modèle (F1={best_f1:.4f})\n")
    else:
        print()

model.load_state_dict(torch.load('best_trimodal_transe_tt.pt'))
print("\n=== ÉVALUATION FINALE (M1+M2+M3 TransE — Titre+Texte) ===\n")
_, acc, f1, auc, preds, true_labels = evaluate(model, test_loader, criterion, device)
print(classification_report(true_labels, preds, target_names=['Réel', 'Fake']))
print(f"Accuracy : {acc:.4f} | F1 : {f1:.4f} | AUC : {auc:.4f}")

# --- Courbes ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(history['train_loss'], label='Train', marker='o')
ax1.plot(history['val_loss'],   label='Val',   marker='o')
ax1.set_title('Loss — M1+M2+M3 TransE (Titre+Texte)') ; ax1.legend()
ax2.plot(history['train_acc'], label='Train', marker='o')
ax2.plot(history['val_acc'],   label='Val',   marker='o')
ax2.set_title('Accuracy — M1+M2+M3 TransE (Titre+Texte)') ; ax2.legend()
plt.tight_layout()
plt.savefig('trimodal_transe_tt_curves.png', dpi=120)
plt.show()

# --- Tableau comparatif ---
print("\n" + "="*65)
print("TABLEAU COMPARATIF — GossipCop (Titre+Texte)")
print("="*65)
print(f"{'Modèle':<42} {'Accuracy':>10} {'F1':>8} {'AUC':>8}")
print("-"*65)
print(f"{'TF-IDF+SVM (titre seul)':<42} {'0.7876':>10} {'0.7875':>8} {'0.8599':>8}")
print(f"{'TF-IDF+SVM (titre+texte)':<42} {'0.8293':>10} {'0.8293':>8} {'0.9066':>8}")
print(f"{'BERT+Source M1+M2 (titre seul)':<42} {'0.8235':>10} {'0.8234':>8} {'0.8942':>8}")
print(f"{'M1+M2+M3 hash (titre+texte)':<42} {'⏳':>10} {'⏳':>8} {'⏳':>8}")
print(f"{'M1+M2+M3 TransE (titre+texte)':<42} {acc:>10.4f} {f1:>8.4f} {auc:>8.4f}")
print("="*65)

torch.save(model.state_dict(), 'trimodal_transe_tt_final.pt')
print("✅ Modèle final sauvegardé dans 'trimodal_transe_tt_final.pt'")
```
